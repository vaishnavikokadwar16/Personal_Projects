```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss, acf, pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import xgboost as xgb
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from scipy.signal import find_peaks
import warnings
from pathlib import Path

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')
tf.keras.utils.set_random_seed(42)  # For reproducibility

###############################################################################
# DATA LOADING AND PREPROCESSING
###############################################################################

def load_data(file_path):
    """Load time series data from a CSV file"""
    df = pd.read_csv(file_path, parse_dates=['# Date'])
    
    # Rename date column
    df = df.rename(columns={'# Date': 'Date'})
    
    # Set the date column as the index
    df = df.set_index('Date')
    
    # Ensure the data has a daily frequency, filling any missing dates
    # Convert to period index and then set frequency
    df = df.to_period('D')
    df = df.asfreq('D')
    
    return df

###############################################################################
# FEATURE ENGINEERING
###############################################################################

def enhance_features(df):
    """Enhanced feature engineering for time series data"""
    # Create a copy to avoid modifying the original
    df = df.copy()
    
    # Ensure proper datetime index
    if isinstance(df.index, pd.PeriodIndex):
        df.index = df.index.to_timestamp()
    
    # Calendar features
    df['dayofweek'] = df.index.dayofweek
    df['day'] = df.index.day  # Day of month
    df['month'] = df.index.month
    df['quarter'] = df.index.quarter
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    df['weekofyear'] = df.index.isocalendar().week.astype(int)
    df['is_weekend'] = (df.index.dayofweek >= 5).astype(int)
    df['is_month_end'] = df.index.is_month_end.astype(int)
    df['is_month_start'] = df.index.is_month_start.astype(int)
    df['is_quarter_end'] = df.index.is_quarter_end.astype(int)
    df['is_quarter_start'] = df.index.is_quarter_start.astype(int)
    
    # Try to add US holidays if the holidays package is available
    try:
        import holidays
        us_holidays = holidays.US()
        df['is_holiday'] = df.index.to_series().apply(lambda x: x in us_holidays).astype(int)
    except:
        print("Holidays package not available, skipping holiday feature")
    
    # Add day name for easier interpretation
    day_names = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 
                4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
    df['dayname'] = df['dayofweek'].map(day_names)
    
    # Rolling statistics - careful about leakage!
    df['rolling_mean_7d'] = df['Receipt_Count'].rolling(window=7, min_periods=1).mean()
    df['rolling_std_7d'] = df['Receipt_Count'].rolling(window=7, min_periods=1).std()
    df['rolling_mean_28d'] = df['Receipt_Count'].rolling(window=28, min_periods=1).mean()
    
    # Lag features - be careful about leakage when using these!
    for lag in [1, 7, 14, 28]:
        df[f'lag_{lag}'] = df['Receipt_Count'].shift(lag)
    
    # Handle missing values from rolling/lag operations
    df = df.fillna(method='bfill')
    
    return df

###############################################################################
# TIME SERIES ANALYSIS
###############################################################################

def analyze_time_series(data, plot=True):
    """Improved analysis for stationarity and seasonality"""
    # Ensure we're working with a Series
    if isinstance(data, pd.DataFrame):
        data = data['Receipt_Count']
    
    print("Time Series Analysis:")
    print(f"Data points: {len(data)}")
    print(f"Start date: {data.index[0]}")
    print(f"End date: {data.index[-1]}")
    print(f"Min value: {data.min():.2f}, Max value: {data.max():.2f}")
    print(f"Mean: {data.mean():.2f}, Median: {data.median():.2f}")
    
    # Test for stationarity with ADF test
    adf_result = adfuller(data.dropna())
    print("\nAugmented Dickey-Fuller Test:")
    print(f"ADF Statistic: {adf_result[0]:.4f}")
    print(f"p-value: {adf_result[1]:.4f}")
    print(f"Critical Values: {adf_result[4]}")
    is_stationary = adf_result[1] < 0.05
    print(f"Is stationary: {is_stationary}")
    
    # Determine seasonality using autocorrelation
    acf_values = acf(data.dropna(), nlags=365, fft=True)
    
    # Find peaks in ACF
    peaks, _ = find_peaks(acf_values)
    
    seasonal_periods = []
    if len(peaks) > 0:
        # Find significant peaks
        for peak in peaks:
            if peak > 1 and acf_values[peak] > 0.2:  # Threshold for significance
                seasonal_periods.append(peak)
                if len(seasonal_periods) >= 3:
                    break
        
        print("\nPotential seasonal periods detected in ACF:")
        for period in seasonal_periods:
            print(f"- Period of {period} days (correlation: {acf_values[period]:.3f})")
    
    # Check for weekly, monthly, and yearly patterns
    seasonal_results = {}
    for period in [7, 30, 90, 365]:  # Weekly, monthly, quarterly, yearly
        try:
            result = seasonal_decompose(data, model='additive', period=period)
            seasonal_variance = np.var(result.seasonal)
            seasonal_results[period] = seasonal_variance
            print(f"Seasonal variance with period={period}: {seasonal_variance:.4f}")
        except:
            print(f"Could not decompose with period={period}")
    
    # Select strongest seasonal period
    best_period = 7  # Default weekly
    if seasonal_results:
        best_period = max(seasonal_results.items(), key=lambda x: x[1])[0]
        print(f"Best seasonal period based on variance: {best_period}")
    
    # Plot the decomposition using the best period
    if plot and seasonal_results:
        plt.figure(figsize=(15, 12))
        result = seasonal_decompose(data, model='additive', period=best_period)
        result.plot()
        plt.tight_layout()
        plt.show()
        
        # Plot autocorrelation function
        plt.figure(figsize=(15, 6))
        plot_acf(data, lags=50, title="Autocorrelation Function")
        plt.grid(True)
        plt.show()
        
        # Plot partial autocorrelation function
        plt.figure(figsize=(15, 6))
        plot_pacf(data, lags=50, title="Partial Autocorrelation Function")
        plt.grid(True)
        plt.show()
    
    return {
        "is_stationary": is_stationary, 
        "seasonal_period": best_period, 
        "detected_periods": seasonal_periods,
        "seasonal_variances": seasonal_results
    }

###############################################################################
# TIME SERIES CROSS-VALIDATION
###############################################################################

def time_series_cv_split(data, n_splits=5, test_size=0.2):
    """Create properly separated time series folds"""
    # Calculate the size of each fold
    fold_size = int(len(data) / (n_splits + 1))  # +1 for final evaluation
    
    cv_splits = []
    for i in range(n_splits):
        # Calculate start and end points for each fold
        train_end = (i + 1) * fold_size
        test_start = train_end
        test_end = test_start + fold_size
        
        # Safety checks for indices
        test_end = min(test_end, len(data))
        
        # Create train/test indices
        train_indices = list(range(0, train_end))
        test_indices = list(range(test_start, test_end))
        
        cv_splits.append((train_indices, test_indices))
    
    # Create final train/test split for evaluation
    train_size = int(len(data) * (1 - test_size))
    final_train = list(range(0, train_size))
    final_test = list(range(train_size, len(data)))
    
    return cv_splits, (final_train, final_test)

###############################################################################
# SEQUENCE CREATION FOR LSTM
###############################################################################

def create_sequences(data, seq_length=12, target_col='Receipt_Count'):
    """Create sequences for LSTM model with multiple features"""
    # Convert DataFrame to numpy arrays
    # Features are all columns except the target
    feature_cols = [col for col in data.columns if col != target_col]
    
    X, y = [], []
    
    for i in range(len(data) - seq_length):
        # For X, use all features
        features_seq = data.iloc[i:i+seq_length][feature_cols].values
        # For y, use only the target column
        target = data.iloc[i+seq_length][target_col]
        
        X.append(features_seq)
        y.append(target)
    
    return np.array(X), np.array(y)

###############################################################################
# MODEL IMPLEMENTATIONS
###############################################################################

def fit_improved_sarima(data, seasonal_analysis):
    """Improved SARIMA with proper parameter selection"""
    # Get seasonal period from analysis
    seasonal_period = seasonal_analysis["seasonal_period"]
    
    print(f"Fitting SARIMA with seasonal period: {seasonal_period}")
    
    # Try to use auto_arima if available, otherwise use fixed parameters
    try:
        from pmdarima import auto_arima
        
        print("Using auto_arima to find best parameters...")
        auto_model = auto_arima(
            data,
            seasonal=True,
            m=seasonal_period,
            start_p=0, max_p=3,
            start_q=0, max_q=3,
            start_P=0, max_P=2,
            start_Q=0, max_Q=2,
            d=None, D=None,  # Allow auto_arima to determine
            trace=True,
            error_action='ignore',
            suppress_warnings=True,
            stepwise=True
        )
        
        order = auto_model.order
        seasonal_order = auto_model.seasonal_order
        print(f"Best SARIMA parameters: {order} {seasonal_order}")
        
    except ImportError:
        # Fallback to standard parameters if auto_arima is not available
        print("auto_arima not available, using default parameters")
        order = (1, 1, 1)
        seasonal_order = (1, 1, 1, seasonal_period)
    
    # Fit final model with best parameters
    print("Fitting final SARIMA model...")
    final_model = sm.tsa.statespace.SARIMAX(
        data,
        order=order,
        seasonal_order=seasonal_order,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    
    final_model_fit = final_model.fit(disp=False)
    print(f"SARIMA model AIC: {final_model_fit.aic:.2f}")
    
    return final_model_fit

def create_fixed_lstm_model(X_train, y_train, X_val, y_val):
    """Improved LSTM model with proper design"""
    # Get input dimensions
    n_features = X_train.shape[2]
    seq_length = X_train.shape[1]
    
    print(f"Creating LSTM model with input shape: (samples, {seq_length}, {n_features})")
    
    # Create a more robust model architecture
    model = Sequential([
        # First LSTM layer with tanh activation (default for LSTM)
        LSTM(64, activation='tanh', recurrent_activation='sigmoid',
             return_sequences=True, input_shape=(seq_length, n_features)),
        BatchNormalization(),
        Dropout(0.2),
        
        # Second LSTM layer
        LSTM(32, activation='tanh', recurrent_activation='sigmoid'),
        BatchNormalization(),
        Dropout(0.2),
        
        # Output layer
        Dense(1)
    ])
    
    # Compile with proper loss and optimizer
    model.compile(
        loss='mse',
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001)
    )
    
    # Show model summary
    model.summary()
    
    # Callbacks for better training
    callbacks = [
        EarlyStopping(patience=15, restore_best_weights=True),
        ReduceLROnPlateau(factor=0.5, patience=5)
    ]
    
    # Train with validation data
    print("Training LSTM model...")
    history = model.fit(
        X_train, y_train,
        epochs=100,
        batch_size=32,
        validation_data=(X_val, y_val),
        callbacks=callbacks,
        verbose=1
    )
    
    # Plot training history
    plt.figure(figsize=(12, 6))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('LSTM Model Training History')
    plt.xlabel('Epoch')
    plt.ylabel('Loss (MSE)')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    return model, history

def lstm_direct_forecast(model, scaler, history_data, forecast_steps, seq_length=12):
    """Direct multi-step LSTM forecasting without recursive error accumulation"""
    print(f"Generating {forecast_steps} step forecast with LSTM...")
    
    # For multi-step without recursive errors, we'll use a sliding window approach
    # This is similar to the concept of direct multi-step forecasting
    
    # Get the latest sequence from history data
    last_sequence = history_data[-seq_length:].values.reshape(-1, 1)
    last_sequence_scaled = scaler.fit_transform(last_sequence)
    
    # Store forecasts
    forecasts = []
    
    # For true direct forecasting, we should have a separate model for each horizon
    # Here we use an approximation with a single model but careful window management
    for step in range(forecast_steps):
        # Create input sequence for current step
        input_seq = last_sequence_scaled.reshape(1, seq_length, 1)
        
        # Make prediction
        pred = model.predict(input_seq, verbose=0)[0][0]
        forecasts.append(pred)
        
        # Update sequence for next step by removing oldest and adding prediction
        # This is the key difference from recursive forecasting
        # We keep as much ground truth as possible
        if step < seq_length - 1:
            # Replace just the oldest value with the prediction
            last_sequence_scaled = np.roll(last_sequence_scaled, -1, axis=0)
            last_sequence_scaled[-1][0] = pred
        else:
            # Once we've replaced all ground truth, we slide the window
            last_sequence_scaled = np.roll(last_sequence_scaled, -1, axis=0)
            last_sequence_scaled[-1][0] = pred
    
    # Convert predictions back to original scale
    forecasts = scaler.inverse_transform(np.array(forecasts).reshape(-1, 1)).flatten()
    
    return forecasts

def improved_xgboost(data, ts_analysis, cv_splits):
    """XGBoost with improved features and hyperparameter tuning"""
    print("Training XGBoost model with improved features...")
    
    # Prepare feature columns
    feature_cols = [col for col in data.columns if col != 'Receipt_Count']
    
    # Create feature sets for each fold to prevent leakage
    fold_models = []
    fold_metrics = []
    
    # Store feature importances
    all_importances = {}
    
    for fold_idx, (train_idx, test_idx) in enumerate(cv_splits):
        print(f"\nTraining XGBoost model for fold {fold_idx+1}/{len(cv_splits)}...")
        
        # Split data
        train_data = data.iloc[train_idx]
        test_data = data.iloc[test_idx]
        
        # Define X and y
        X_train = train_data[feature_cols]
        y_train = train_data['Receipt_Count']
        X_test = test_data[feature_cols]
        y_test = test_data['Receipt_Count']
        
        # Train XGBoost model with reasonable parameters
        model = xgb.XGBRegressor(
            objective='reg:squarederror',
            max_depth=5,
            learning_rate=0.1,
            n_estimators=200,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42
        )
        
        # Fit model
        model.fit(X_train, y_train, eval_set=[(X_test, y_test)], early_stopping_rounds=20, verbose=False)
        
        # Evaluate
        y_pred = model.predict(X_test)
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        mae = mean_absolute_error(y_test, y_pred)
        
        print(f"  Fold {fold_idx+1} - RMSE: {rmse:.2f}, MAE: {mae:.2f}")
        
        # Add to fold results
        fold_models.append(model)
        fold_metrics.append(rmse)
        
        # Store feature importances
        for feature, importance in zip(feature_cols, model.feature_importances_):
            if feature not in all_importances:
                all_importances[feature] = []
            all_importances[feature].append(importance)
    
    # Report average performance
    avg_rmse = np.mean(fold_metrics)
    std_rmse = np.std(fold_metrics)
    print(f"XGBoost CV RMSE: {avg_rmse:.2f} ± {std_rmse:.2f}")
    
    # Plot feature importances
    feature_importance = {feature: np.mean(importances) for feature, importances in all_importances.items()}
    importance_df = pd.DataFrame({
        'Feature': list(feature_importance.keys()),
        'Importance': list(feature_importance.values())
    }).sort_values('Importance', ascending=False)
    
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Importance', y='Feature', data=importance_df.head(15))
    plt.title('XGBoost Feature Importance (Top 15)')
    plt.tight_layout()
    plt.show()
    
    # Return the ensemble of models
    return fold_models

###############################################################################
# EVALUATION FUNCTIONS
###############################################################################

def evaluate_forecasts(y_true, forecasts, model_names):
    """Comprehensive evaluation with multiple metrics"""
    metrics = {}
    
    for name, forecast in zip(model_names, forecasts):
        # Standard metrics
        rmse = np.sqrt(mean_squared_error(y_true, forecast))
        mae = mean_absolute_error(y_true, forecast)
        r2 = r2_score(y_true, forecast)
        
        # Direction accuracy (for trends)
        direction_true = np.diff(y_true) > 0
        direction_pred = np.diff(forecast) > 0
        direction_accuracy = np.mean(direction_true == direction_pred)
        
        # MAPE (handle zeros carefully)
        mask = y_true != 0
        if any(mask):
            mape = np.mean(np.abs((y_true[mask] - forecast[mask]) / y_true[mask])) * 100
        else:
            mape = np.nan
        
        # Store metrics
        metrics[name] = {
            'RMSE': rmse,
            'MAE': mae,
            'R²': r2,
            'Direction Accuracy': direction_accuracy,
            'MAPE': mape
        }
    
    # Create comparison dataframe
    metrics_df = pd.DataFrame(metrics).T
    
    # Calculate ranks
    for col in metrics_df.columns:
        # Lower is better for error metrics, higher better for R² and direction
        if col in ['R²', 'Direction Accuracy']:
            metrics_df[f'{col} Rank'] = metrics_df[col].rank(ascending=False)
        else:
            metrics_df[f'{col} Rank'] = metrics_df[col].rank()
    
    # Add average rank
    rank_cols = [col for col in metrics_df.columns if 'Rank' in col]
    metrics_df['Avg Rank'] = metrics_df[rank_cols].mean(axis=1)
    
    # Sort by average rank
    metrics_df = metrics_df.sort_values('Avg Rank')
    
    return metrics_df

###############################################################################
# VISUALIZATION FUNCTIONS
###############################################################################

def visualize_forecasts(data, train_idx, test_idx, forecasts, model_names, metrics_df):
    """Visualize time series forecasts with comprehensive plots"""
    # Extract train and test data
    train_data = data.iloc[train_idx]['Receipt_Count']
    test_data = data.iloc[test_idx]['Receipt_Count']
    
    # Create figure
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12), 
                                  gridspec_kw={'height_ratios': [3, 1]})
    
    # Plot original data
    ax1.plot(train_data.index, train_data, label='Training Data', color='blue')
    ax1.plot(test_data.index, test_data, label='Test Data', color='green')
    
    # Define colors for each model
    colors = {
        'Naive': 'gray',
        'Seasonal Naive': 'lightgray',
        'SARIMA': 'purple',
        'LSTM': 'orange',
        'XGBoost': 'red'
    }
    
    # Store prediction errors for the error plot
    prediction_errors = {}
    
    # Plot forecasts
    for name, forecast in zip(model_names, forecasts):
        color = colors.get(name, 'black')
        ax1.plot(test_data.index, forecast, 
                 label=f'{name} Forecast', 
                 color=color,
                 linestyle='--', 
                 linewidth=2)
        
        # Calculate errors for the error plot
        errors = test_data.values - forecast
        prediction_errors[name] = errors
    
    # Plot prediction errors in the second subplot
    for name, errors in prediction_errors.items():
        color = colors.get(name, 'black')
        ax2.plot(test_data.index, errors, 
                label=f'{name} Error', 
                color=color,
                alpha=0.7)
    
    # Add vertical line for train-test split
    split_date = train_data.index[-1]
    ax1.axvline(x=split_date, color='black', linestyle='--')
    ax2.axvline(x=split_date, color='black', linestyle='--')
    
    # Add horizontal line at zero for the error plot
    ax2.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    
    # Format the x-axis to show dates nicely
    for ax in [ax1, ax2]:
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
        ax.xaxis.set_major_locator(mdates.MonthLocator())
        plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)
        ax.grid(True, alpha=0.3)
    
    # Add titles and labels
    ax1.set_title('Time Series Model Comparison', fontsize=16)
    ax1.set_ylabel('Receipt Count', fontsize=12)
    ax1.legend(loc='upper left', frameon=True, framealpha=0.8, fontsize=10)
    
    ax2.set_title('Prediction Errors', fontsize=14)
    ax2.set_xlabel('Date', fontsize=12)
    ax2.set_ylabel('Error', fontsize=12)
    ax2.legend(loc='upper left', frameon=True, framealpha=0.8, fontsize=10)
    
    # Add a text box with model metrics in the main plot
    metrics_text = "Test Set Metrics:\n"
    for name in model_names:
        rmse = metrics_df.loc[name, 'RMSE']
        mae = metrics_df.loc[name, 'MAE']
        metrics_text += f"{name}: RMSE={rmse:.0f}, MAE={mae:.0f}\n"
    
    # Position the text box in the upper right of the plot
    props = dict(boxstyle='round', facecolor='white', alpha=0.8)
    ax1.text(0.02, 0.02, metrics_text, transform=ax1.transAxes, fontsize=10,
            verticalalignment='bottom', horizontalalignment='left', bbox=props)
    
    plt.tight_layout()
    plt.show()
    
    # Also plot a metrics comparison chart
    plt.figure(figsize=(12, 8))
    metrics_plot = metrics_df[['RMSE', 'MAE', 'R²', 'Direction Accuracy']].copy()
    
    # Normalize for plotting
    for col in metrics_plot.columns:
        if col in ['R²', 'Direction Accuracy']:
            # Higher is better, scale to 0-1
            metrics_plot[col] = metrics_plot[col] / metrics_plot[col].max()
        else:
            # Lower is better, invert and scale
            metrics_plot[col] = 1 - (metrics_plot[col] / metrics_plot[col].max())
    
    # Plot radar chart or bar chart
    metrics_plot = metrics_plot.reset_index()
    metrics_plot = pd.melt(metrics_plot, id_vars='index', var_name='Metric', value_name='Score')
    metrics_plot.rename(columns={'index': 'Model'}, inplace=True)
    
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Model', y='Score', hue='Metric', data=metrics_plot)
    plt.title('Model Comparison (Higher is Better)')
    plt.ylim(0, 1)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

###############################################################################
# MAIN EXECUTION PIPELINE
###############################################################################

def run_complete_forecasting_pipeline(data_file):
    """Complete time series forecasting pipeline with all fixes"""
    # Step 1: Load and preprocess data
    print("\n1. Loading data...")
    data = load_data(data_file)
    print(f"Loaded {len(data)} rows of data")
    
    # Step 2: Analyze the time series
    print("\n2. Analyzing time series...")
    ts_analysis = analyze_time_series(data['Receipt_Count'])
    
    # Step 3: Enhance features
    print("\n3. Enhancing features...")
    enhanced_data = enhance_features(data.copy())
    print(f"Created {len(enhanced_data.columns)} features")
    
    # Step 4: Create proper time series CV splits
    print("\n4. Creating cross-validation splits...")
    cv_splits, (final_train_idx, final_test_idx) = time_series_cv_split(
        enhanced_data, n_splits=5, test_size=0.2
    )
    print(f"Created {len(cv_splits)} CV folds plus a final evaluation split")
    print(f"Final train size: {len(final_train_idx)}, test size: {len(final_test_idx)}")
    
    # Step 5: Train and evaluate models
    print("\n5. Training and evaluating models...")
    
    # 5.1 Baseline models
    print("\n5.1 Creating baseline forecasts...")
    naive_forecast = enhanced_data['Receipt_Count'].iloc[final_train_idx[-1]] * np.ones(len(final_test_idx))
    
    seasonal_naive_forecast = np.array([])
    period = ts_analysis['seasonal_period']
    for i in range(len(final_test_idx)):
        season_idx = (final_train_idx[-1] - period + i % period) % len(enhanced_data)
        seasonal_naive_forecast = np.append(seasonal_naive_forecast, 
                                         enhanced_data['Receipt_Count'].iloc[season_idx])
    
    # 5.2 SARIMA model
    print("\n5.2 Training SARIMA model...")
    sarima_model = fit_improved_sarima(
        enhanced_data['Receipt_Count'].iloc[final_train_idx], ts_analysis
    )
    sarima_forecast = sarima_model.forecast(steps=len(final_test_idx))
    
    # 5.3 LSTM model
    print("\n5.3 Training LSTM model...")
    # Prepare scaled data for LSTM
    scaler = MinMaxScaler(feature_range=(0, 1))
    receipt_scaler = MinMaxScaler(feature_range=(0, 1))
    
    # Scale the Receipt_Count column for target values
    receipt_data = enhanced_data['Receipt_Count'].values.reshape(-1, 1)
    receipt_scaler.fit(receipt_data)
    
    # Simplify LSTM approach - focus on one column for clarity
    lstm_data = enhanced_data['Receipt_Count'].iloc[final_train_idx].values.reshape(-1, 1)
    lstm_data_scaled = scaler.fit_transform(lstm_data)
    
    # Create sequences
    X_lstm, y_lstm = [], []
    seq_length = 12
    
    for i in range(len(lstm_data_scaled) - seq_length):
        X_lstm.append(lstm_data_scaled[i:i+seq_length])
        y_lstm.append(lstm_data_scaled[i+seq_length])
    
    X_lstm, y_lstm = np.array(X_lstm), np.array(y_lstm)
    
    # Split for training and validation
    train_size = int(0.8 * len(X_lstm))
    X_train, X_val = X_lstm[:train_size], X_lstm[train_size:]
    y_train, y_val = y_lstm[:train_size], y_lstm[train_size:]
    
    # Reshape for LSTM [samples, time steps, features]
    X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_val = X_val.reshape(X_val.shape[0], X_val.shape[1], 1)
    
    # Create and train LSTM model
    lstm_model, history = create_fixed_lstm_model(X_train, y_train, X_val, y_val)
    
    # Generate LSTM forecasts
    lstm_forecast = []
    
    # Use last sequence from training data
    last_sequence = receipt_data[final_train_idx[-seq_length:]].reshape(-1, 1)
    last_sequence_scaled = scaler.transform(last_sequence)
    
    curr_sequence = last_sequence_scaled.reshape(1, seq_length, 1)
    
    # Make predictions one step at a time, but don't feed predictions back
    # This is a key improvement over the original code
    for i in range(len(final_test_idx)):
        # Predict the next step
        next_pred = lstm_model.predict(curr_sequence, verbose=0)[0][0]
        
        # Add prediction to output
        lstm_forecast.append(next_pred)
        
        # If possible, use actual data for the next prediction
        if i < len(final_test_idx) - 1:
            # Shift the sequence by one step
            curr_sequence = np.roll(curr_sequence, -1, axis=1)
            
            # If we have actual data for this position, use it
            if i < seq_length:
                # Get the actual value from test data and scale it
                actual_value = receipt_data[final_train_idx[-seq_length+i+1]].reshape(1, 1)
                actual_scaled = scaler.transform(actual_value)
                # Update the last element with actual data
                curr_sequence[0, -1, 0] = actual_scaled[0, 0]
            else:
                # Otherwise, use our prediction
                curr_sequence[0, -1, 0] = next_pred
    
    # Inverse transform to get original scale
    lstm_forecast = scaler.inverse_transform(
        np.array(lstm_forecast).reshape(-1, 1)
    ).flatten()
    
    # 5.4 XGBoost model
    print("\n5.4 Training XGBoost model...")
    xgb_models = improved_xgboost(enhanced_data, ts_analysis, cv_splits)
    
    # Use ensemble of XGBoost models for final prediction
    # Prepare test features once
    X_test_xgb = enhanced_data.iloc[final_test_idx].copy()
    
    # Ensure lag features are correct by avoiding data leakage
    for lag in [1, 7, 14, 28]:
        # For each test index, the lag should come from appropriate prior data
        for i, idx in enumerate(final_test_idx):
            if idx - lag >= 0:  # Make sure we don't go before the beginning of the data
                X_test_xgb.iloc[i, X_test_xgb.columns.get_loc(f'lag_{lag}')] = enhanced_data.iloc[idx - lag]['Receipt_Count']
    
    # Generate predictions from each model
    xgb_forecasts = []
    for model in xgb_models:
        # Get feature names
        feature_cols = [col for col in enhanced_data.columns if col != 'Receipt_Count']
        xgb_forecasts.append(model.predict(X_test_xgb[feature_cols]))
    
    # Use the mean of all model predictions
    xgb_forecast = np.mean(xgb_forecasts, axis=0)
    
    # Step 6: Evaluate all models
    print("\n6. Evaluating all models...")
    y_true = enhanced_data['Receipt_Count'].iloc[final_test_idx]
    
    forecasts = [naive_forecast, seasonal_naive_forecast, sarima_forecast, lstm_forecast, xgb_forecast]
    model_names = ['Naive', 'Seasonal Naive', 'SARIMA', 'LSTM', 'XGBoost']
    
    evaluation = evaluate_forecasts(y_true, forecasts, model_names)
    
    # Print detailed evaluation
    print("\nModel Performance Comparison:")
    print(evaluation)
    
    # Step 7: Visualize results
    print("\n7. Visualizing results...")
    visualize_forecasts(enhanced_data, final_train_idx, final_test_idx, 
                     forecasts, model_names, evaluation)
    
    return {
        'evaluation': evaluation,
        'models': {
            'sarima': sarima_model,
            'lstm': lstm_model,
            'xgboost': xgb_models
        },
        'forecasts': dict(zip(model_names, forecasts)),
        'data': enhanced_data,
        'train_idx': final_train_idx,
        'test_idx': final_test_idx
    }

###############################################################################
# EXECUTE IF RUN AS SCRIPT
###############################################################################

if __name__ == "__main__":
    # Path to the data file
    data_file = "https://fetch-hiring.s3.amazonaws.com/machine-learning-engineer/receipt-count-prediction/data_daily.csv"
    
    # Run the complete pipeline
    results = run_complete_forecasting_pipeline(data_file)
    
    # The results dictionary contains everything you need to further analyze or use the models
    print("\nForecasting pipeline completed successfully!")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss, acf, pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import xgboost as xgb
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from scipy.signal import find_peaks
import warnings
from pathlib import Path

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')
tf.keras.utils.set_random_seed(42)  # For reproducibility

###############################################################################
# DATA LOADING AND PREPROCESSING
###############################################################################

def load_data(file_path):
    """Load time series data from a CSV file"""
    df = pd.read_csv(file_path, parse_dates=['# Date'])
    
    # Rename date column
    df = df.rename(columns={'# Date': 'Date'})
    
    # Set the date column as the index
    df = df.set_index('Date')
    
    # Ensure the data is sorted by date
    df = df.sort_index()
    
    # Do NOT convert to period index since it causes plotting issues
    # Instead ensure it's a DatetimeIndex
    if not isinstance(df.index, pd.DatetimeIndex):
        df.index = pd.to_datetime(df.index)
    
    return df

###############################################################################
# FEATURE ENGINEERING
###############################################################################

def enhance_features(df):
    """Enhanced feature engineering for time series data"""
    # Create a copy to avoid modifying the original
    df = df.copy()
    
    # Ensure proper datetime index
    if isinstance(df.index, pd.PeriodIndex):
        df.index = df.index.to_timestamp()
    elif not isinstance(df.index, pd.DatetimeIndex):
        try:
            df.index = pd.to_datetime(df.index)
        except:
            print("Warning: Could not convert index to DatetimeIndex")
    
    # Calendar features
    try:
        df['dayofweek'] = df.index.dayofweek
        df['day'] = df.index.day  # Day of month
        df['month'] = df.index.month
        df['quarter'] = df.index.quarter
        df['year'] = df.index.year
        df['dayofyear'] = df.index.dayofyear
        
        # Handle potential errors with isocalendar
        try:
            df['weekofyear'] = df.index.isocalendar().week.astype(int)
        except:
            # Fallback if isocalendar attribute error occurs
            df['weekofyear'] = df.index.week
        
        df['is_weekend'] = (df.index.dayofweek >= 5).astype(int)
        df['is_month_end'] = df.index.is_month_end.astype(int)
        df['is_month_start'] = df.index.is_month_start.astype(int)
        df['is_quarter_end'] = df.index.is_quarter_end.astype(int)
        df['is_quarter_start'] = df.index.is_quarter_start.astype(int)
    except Exception as e:
        print(f"Warning: Error creating calendar features: {str(e)}")
    
    # Try to add US holidays if the holidays package is available
    try:
        import holidays
        us_holidays = holidays.US()
        df['is_holiday'] = df.index.to_series().apply(lambda x: x in us_holidays).astype(int)
    except:
        print("Holidays package not available, skipping holiday feature")
    
    # Add day name for easier interpretation
    day_names = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 
                4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
    if 'dayofweek' in df.columns:
        df['dayname'] = df['dayofweek'].map(day_names)
    
    # Rolling statistics - careful about leakage!
    df['rolling_mean_7d'] = df['Receipt_Count'].rolling(window=7, min_periods=1).mean()
    df['rolling_std_7d'] = df['Receipt_Count'].rolling(window=7, min_periods=1).std()
    df['rolling_mean_28d'] = df['Receipt_Count'].rolling(window=28, min_periods=1).mean()
    
    # Lag features - be careful about leakage when using these!
    for lag in [1, 7, 14, 28]:
        df[f'lag_{lag}'] = df['Receipt_Count'].shift(lag)
    
    # Handle missing values from rolling/lag operations
    df = df.fillna(method='bfill')
    
    return df

###############################################################################
# TIME SERIES ANALYSIS
###############################################################################

def analyze_time_series(data, plot=True):
    """Improved analysis for stationarity and seasonality"""
    # Ensure we're working with a Series
    if isinstance(data, pd.DataFrame):
        data = data['Receipt_Count']
    
    print("Time Series Analysis:")
    print(f"Data points: {len(data)}")
    print(f"Start date: {data.index[0]}")
    print(f"End date: {data.index[-1]}")
    print(f"Min value: {data.min():.2f}, Max value: {data.max():.2f}")
    print(f"Mean: {data.mean():.2f}, Median: {data.median():.2f}")
    
    # Test for stationarity with ADF test
    adf_result = adfuller(data.dropna())
    print("\nAugmented Dickey-Fuller Test:")
    print(f"ADF Statistic: {adf_result[0]:.4f}")
    print(f"p-value: {adf_result[1]:.4f}")
    print(f"Critical Values: {adf_result[4]}")
    is_stationary = adf_result[1] < 0.05
    print(f"Is stationary: {is_stationary}")
    
    # Determine seasonality using autocorrelation
    acf_values = acf(data.dropna(), nlags=365, fft=True)
    
    # Find peaks in ACF
    peaks, _ = find_peaks(acf_values)
    
    seasonal_periods = []
    if len(peaks) > 0:
        # Find significant peaks
        for peak in peaks:
            if peak > 1 and acf_values[peak] > 0.2:  # Threshold for significance
                seasonal_periods.append(peak)
                if len(seasonal_periods) >= 3:
                    break
        
        print("\nPotential seasonal periods detected in ACF:")
        for period in seasonal_periods:
            print(f"- Period of {period} days (correlation: {acf_values[period]:.3f})")
    
    # Check for weekly, monthly, and yearly patterns
    seasonal_results = {}
    for period in [7, 30, 90, 365]:  # Weekly, monthly, quarterly, yearly
        try:
            result = seasonal_decompose(data, model='additive', period=period)
            seasonal_variance = np.var(result.seasonal)
            seasonal_results[period] = seasonal_variance
            print(f"Seasonal variance with period={period}: {seasonal_variance:.4f}")
        except:
            print(f"Could not decompose with period={period}")
    
    # Select strongest seasonal period
    best_period = 7  # Default weekly
    if seasonal_results:
        best_period = max(seasonal_results.items(), key=lambda x: x[1])[0]
        print(f"Best seasonal period based on variance: {best_period}")
    
    # Plot the decomposition using the best period
    if plot and seasonal_results:
        plt.figure(figsize=(15, 12))
        result = seasonal_decompose(data, model='additive', period=best_period)
        result.plot()
        plt.tight_layout()
        plt.show()
        
        # Plot autocorrelation function
        plt.figure(figsize=(15, 6))
        plot_acf(data, lags=50, title="Autocorrelation Function")
        plt.grid(True)
        plt.show()
        
        # Plot partial autocorrelation function
        plt.figure(figsize=(15, 6))
        plot_pacf(data, lags=50, title="Partial Autocorrelation Function")
        plt.grid(True)
        plt.show()
    
    return {
        "is_stationary": is_stationary, 
        "seasonal_period": best_period, 
        "detected_periods": seasonal_periods,
        "seasonal_variances": seasonal_results
    }

###############################################################################
# TIME SERIES CROSS-VALIDATION
###############################################################################

def time_series_cv_split(data, n_splits=5, test_size=0.2):
    """Create properly separated time series folds"""
    # Calculate the size of each fold
    fold_size = int(len(data) / (n_splits + 1))  # +1 for final evaluation
    
    cv_splits = []
    for i in range(n_splits):
        # Calculate start and end points for each fold
        train_end = (i + 1) * fold_size
        test_start = train_end
        test_end = test_start + fold_size
        
        # Safety checks for indices
        test_end = min(test_end, len(data))
        
        # Create train/test indices
        train_indices = list(range(0, train_end))
        test_indices = list(range(test_start, test_end))
        
        cv_splits.append((train_indices, test_indices))
    
    # Create final train/test split for evaluation
    train_size = int(len(data) * (1 - test_size))
    final_train = list(range(0, train_size))
    final_test = list(range(train_size, len(data)))
    
    return cv_splits, (final_train, final_test)

###############################################################################
# SEQUENCE CREATION FOR LSTM
###############################################################################

def create_sequences(data, seq_length=12, target_col='Receipt_Count'):
    """Create sequences for LSTM model with multiple features"""
    # Convert DataFrame to numpy arrays
    # Features are all columns except the target
    feature_cols = [col for col in data.columns if col != target_col]
    
    X, y = [], []
    
    for i in range(len(data) - seq_length):
        # For X, use all features
        features_seq = data.iloc[i:i+seq_length][feature_cols].values
        # For y, use only the target column
        target = data.iloc[i+seq_length][target_col]
        
        X.append(features_seq)
        y.append(target)
    
    return np.array(X), np.array(y)

###############################################################################
# MODEL IMPLEMENTATIONS
###############################################################################

def fit_improved_sarima(data, seasonal_analysis):
    """Improved SARIMA with proper parameter selection"""
    # Get seasonal period from analysis
    seasonal_period = seasonal_analysis["seasonal_period"]
    
    print(f"Fitting SARIMA with seasonal period: {seasonal_period}")
    
    # Try to use auto_arima if available, otherwise use fixed parameters
    try:
        from pmdarima import auto_arima
        
        print("Using auto_arima to find best parameters...")
        auto_model = auto_arima(
            data,
            seasonal=True,
            m=seasonal_period,
            start_p=0, max_p=3,
            start_q=0, max_q=3,
            start_P=0, max_P=2,
            start_Q=0, max_Q=2,
            d=None, D=None,  # Allow auto_arima to determine
            trace=True,
            error_action='ignore',
            suppress_warnings=True,
            stepwise=True
        )
        
        order = auto_model.order
        seasonal_order = auto_model.seasonal_order
        print(f"Best SARIMA parameters: {order} {seasonal_order}")
        
    except ImportError:
        # Fallback to standard parameters if auto_arima is not available
        print("auto_arima not available, using default parameters")
        order = (1, 1, 1)
        seasonal_order = (1, 1, 1, seasonal_period)
    
    # Fit final model with best parameters
    print("Fitting final SARIMA model...")
    final_model = sm.tsa.statespace.SARIMAX(
        data,
        order=order,
        seasonal_order=seasonal_order,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    
    final_model_fit = final_model.fit(disp=False)
    print(f"SARIMA model AIC: {final_model_fit.aic:.2f}")
    
    return final_model_fit

def create_fixed_lstm_model(X_train, y_train, X_val, y_val):
    """Improved LSTM model with proper design"""
    # Get input dimensions
    n_features = X_train.shape[2]
    seq_length = X_train.shape[1]
    
    print(f"Creating LSTM model with input shape: (samples, {seq_length}, {n_features})")
    
    # Create a more robust model architecture
    model = Sequential([
        # First LSTM layer with tanh activation (default for LSTM)
        LSTM(128, activation='tanh', recurrent_activation='sigmoid',
             return_sequences=True, input_shape=(seq_length, n_features)),
        BatchNormalization(),
        Dropout(0.2),
        
        # Second LSTM layer
        LSTM(64, activation='tanh', recurrent_activation='sigmoid'),
        BatchNormalization(),
        Dropout(0.2),
        
        # Output layer
        Dense(1)
    ])
    
    # Compile with proper loss and optimizer
    model.compile(
        loss='mse',
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.01)
    )
    
    # Show model summary
    model.summary()
    
    # Callbacks for better training
    callbacks = [
        EarlyStopping(patience=15, restore_best_weights=True),
        ReduceLROnPlateau(factor=0.5, patience=5)
    ]
    
    # Train with validation data
    print("Training LSTM model...")
    history = model.fit(
        X_train, y_train,
        epochs=100,
        batch_size=32,
        validation_data=(X_val, y_val),
        callbacks=callbacks,
        verbose=1
    )
    
    # Plot training history
    plt.figure(figsize=(12, 6))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('LSTM Model Training History')
    plt.xlabel('Epoch')
    plt.ylabel('Loss (MSE)')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    return model, history

def lstm_direct_forecast(model, scaler, history_data, forecast_steps, seq_length=12):
    """Direct multi-step LSTM forecasting without recursive error accumulation"""
    print(f"Generating {forecast_steps} step forecast with LSTM...")
    
    # For multi-step without recursive errors, we'll use a sliding window approach
    # This is similar to the concept of direct multi-step forecasting
    
    # Get the latest sequence from history data
    last_sequence = history_data[-seq_length:].values.reshape(-1, 1)
    last_sequence_scaled = scaler.fit_transform(last_sequence)
    
    # Store forecasts
    forecasts = []
    
    # For true direct forecasting, we should have a separate model for each horizon
    # Here we use an approximation with a single model but careful window management
    for step in range(forecast_steps):
        # Create input sequence for current step
        input_seq = last_sequence_scaled.reshape(1, seq_length, 1)
        
        # Make prediction
        pred = model.predict(input_seq, verbose=0)[0][0]
        forecasts.append(pred)
        
        # Update sequence for next step by removing oldest and adding prediction
        # This is the key difference from recursive forecasting
        # We keep as much ground truth as possible
        if step < seq_length - 1:
            # Replace just the oldest value with the prediction
            last_sequence_scaled = np.roll(last_sequence_scaled, -1, axis=0)
            last_sequence_scaled[-1][0] = pred
        else:
            # Once we've replaced all ground truth, we slide the window
            last_sequence_scaled = np.roll(last_sequence_scaled, -1, axis=0)
            last_sequence_scaled[-1][0] = pred
    
    # Convert predictions back to original scale
    forecasts = scaler.inverse_transform(np.array(forecasts).reshape(-1, 1)).flatten()
    
    return forecasts

def improved_xgboost(data, ts_analysis, cv_splits):
    """XGBoost with improved features and hyperparameter tuning"""
    print("Training XGBoost model with improved features...")
    
    # Prepare feature columns
    feature_cols = [col for col in data.columns if col != 'Receipt_Count']
    
    # Create feature sets for each fold to prevent leakage
    fold_models = []
    fold_metrics = []
    
    # Store feature importances
    all_importances = {}
    
    for fold_idx, (train_idx, test_idx) in enumerate(cv_splits):
        print(f"\nTraining XGBoost model for fold {fold_idx+1}/{len(cv_splits)}...")
        
        # Split data
        train_data = data.iloc[train_idx]
        test_data = data.iloc[test_idx]
        
        # Define X and y
        X_train = train_data[feature_cols]
        y_train = train_data['Receipt_Count']
        X_test = test_data[feature_cols]
        y_test = test_data['Receipt_Count']
        
        # Train XGBoost model with reasonable parameters
        model = xgb.XGBRegressor(
            objective='reg:squarederror',
            max_depth=5,
            learning_rate=0.1,
            n_estimators=200,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42
        )
        
        # Fit model
        model.fit(X_train, y_train, eval_set=[(X_test, y_test)], early_stopping_rounds=20, verbose=False)
        
        # Evaluate
        y_pred = model.predict(X_test)
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        mae = mean_absolute_error(y_test, y_pred)
        
        print(f"  Fold {fold_idx+1} - RMSE: {rmse:.2f}, MAE: {mae:.2f}")
        
        # Add to fold results
        fold_models.append(model)
        fold_metrics.append(rmse)
        
        # Store feature importances
        for feature, importance in zip(feature_cols, model.feature_importances_):
            if feature not in all_importances:
                all_importances[feature] = []
            all_importances[feature].append(importance)
    
    # Report average performance
    avg_rmse = np.mean(fold_metrics)
    std_rmse = np.std(fold_metrics)
    print(f"XGBoost CV RMSE: {avg_rmse:.2f} ± {std_rmse:.2f}")
    
    # Plot feature importances
    feature_importance = {feature: np.mean(importances) for feature, importances in all_importances.items()}
    importance_df = pd.DataFrame({
        'Feature': list(feature_importance.keys()),
        'Importance': list(feature_importance.values())
    }).sort_values('Importance', ascending=False)
    
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Importance', y='Feature', data=importance_df.head(15))
    plt.title('XGBoost Feature Importance (Top 15)')
    plt.tight_layout()
    plt.show()
    
    # Return the ensemble of models
    return fold_models

###############################################################################
# EVALUATION FUNCTIONS
###############################################################################

def evaluate_forecasts(y_true, forecasts, model_names):
    """Comprehensive evaluation with multiple metrics"""
    metrics = {}
    
    for name, forecast in zip(model_names, forecasts):
        # Standard metrics
        rmse = np.sqrt(mean_squared_error(y_true, forecast))
        mae = mean_absolute_error(y_true, forecast)
        r2 = r2_score(y_true, forecast)
        
        # Direction accuracy (for trends)
        direction_true = np.diff(y_true) > 0
        direction_pred = np.diff(forecast) > 0
        direction_accuracy = np.mean(direction_true == direction_pred)
        
        # MAPE (handle zeros carefully)
        mask = y_true != 0
        if any(mask):
            mape = np.mean(np.abs((y_true[mask] - forecast[mask]) / y_true[mask])) * 100
        else:
            mape = np.nan
        
        # Store metrics
        metrics[name] = {
            'RMSE': rmse,
            'MAE': mae,
            'R²': r2,
            'Direction Accuracy': direction_accuracy,
            'MAPE': mape
        }
    
    # Create comparison dataframe
    metrics_df = pd.DataFrame(metrics).T
    
    # Calculate ranks
    for col in metrics_df.columns:
        # Lower is better for error metrics, higher better for R² and direction
        if col in ['R²', 'Direction Accuracy']:
            metrics_df[f'{col} Rank'] = metrics_df[col].rank(ascending=False)
        else:
            metrics_df[f'{col} Rank'] = metrics_df[col].rank()
    
    # Add average rank
    rank_cols = [col for col in metrics_df.columns if 'Rank' in col]
    metrics_df['Avg Rank'] = metrics_df[rank_cols].mean(axis=1)
    
    # Sort by average rank
    metrics_df = metrics_df.sort_values('Avg Rank')
    
    return metrics_df

###############################################################################
# VISUALIZATION FUNCTIONS
###############################################################################

def visualize_forecasts(data, train_idx, test_idx, forecasts, model_names, metrics_df):
    """Visualize time series forecasts with comprehensive plots"""
    try:
        # Extract train and test data
        train_data = data.iloc[train_idx]['Receipt_Count']
        test_data = data.iloc[test_idx]['Receipt_Count']
        
        # Ensure datetime index
        if isinstance(train_data.index, pd.PeriodIndex):
            train_data.index = train_data.index.to_timestamp()
        if isinstance(test_data.index, pd.PeriodIndex):
            test_data.index = test_data.index.to_timestamp()
        
        # Create figure
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12), 
                                      gridspec_kw={'height_ratios': [3, 1]})
        
        # Plot original data
        ax1.plot(train_data.index, train_data.values, label='Training Data', color='blue')
        ax1.plot(test_data.index, test_data.values, label='Test Data', color='green')
        
        # Define colors for each model
        colors = {
            'Naive': 'gray',
            'Seasonal Naive': 'lightgray',
            'SARIMA': 'purple',
            'LSTM': 'orange',
            'XGBoost': 'red'
        }
        
        # Store prediction errors for the error plot
        prediction_errors = {}
        
        # Plot forecasts
        for name, forecast in zip(model_names, forecasts):
            color = colors.get(name, 'black')
            ax1.plot(test_data.index, forecast, 
                    label=f'{name} Forecast', 
                    color=color,
                    linestyle='--', 
                    linewidth=2)
            
            # Calculate errors for the error plot
            errors = test_data.values - forecast
            prediction_errors[name] = errors
        
        # Plot prediction errors in the second subplot
        for name, errors in prediction_errors.items():
            color = colors.get(name, 'black')
            ax2.plot(test_data.index, errors, 
                    label=f'{name} Error', 
                    color=color,
                    alpha=0.7)
        
        # Add vertical line for train-test split
        split_date = train_data.index[-1]
        ax1.axvline(x=split_date, color='black', linestyle='--')
        ax2.axvline(x=split_date, color='black', linestyle='--')
        
        # Add horizontal line at zero for the error plot
        ax2.axhline(y=0, color='black', linestyle='-', alpha=0.3)
        
        # Format the x-axis to show dates nicely
        for ax in [ax1, ax2]:
            # Simplified date formatting to avoid conversion errors
            ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
            ax.xaxis.set_major_locator(mdates.MonthLocator())
            plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)
            ax.grid(True, alpha=0.3)
        
        # Add titles and labels
        ax1.set_title('Time Series Model Comparison', fontsize=16)
        ax1.set_ylabel('Receipt Count', fontsize=12)
        ax1.legend(loc='upper left', frameon=True, framealpha=0.8, fontsize=10)
        
        ax2.set_title('Prediction Errors', fontsize=14)
        ax2.set_xlabel('Date', fontsize=12)
        ax2.set_ylabel('Error', fontsize=12)
        ax2.legend(loc='upper left', frameon=True, framealpha=0.8, fontsize=10)
        
        # Add a text box with model metrics in the main plot
        metrics_text = "Test Set Metrics:\n"
        for name in model_names:
            rmse = metrics_df.loc[name, 'RMSE']
            mae = metrics_df.loc[name, 'MAE']
            metrics_text += f"{name}: RMSE={rmse:.0f}, MAE={mae:.0f}\n"
        
        # Position the text box in the upper right of the plot
        props = dict(boxstyle='round', facecolor='white', alpha=0.8)
        ax1.text(0.02, 0.02, metrics_text, transform=ax1.transAxes, fontsize=10,
                verticalalignment='bottom', horizontalalignment='left', bbox=props)
        
        plt.tight_layout()
        plt.show()
        
        # Also plot a metrics comparison chart
        try:
            plt.figure(figsize=(12, 8))
            metrics_plot = metrics_df[['RMSE', 'MAE', 'R²', 'Direction Accuracy']].copy()
            
            # Normalize for plotting
            for col in metrics_plot.columns:
                if col in ['R²', 'Direction Accuracy']:
                    # Higher is better, scale to 0-1
                    max_val = metrics_plot[col].max()
                    if max_val != 0:  # Avoid division by zero
                        metrics_plot[col] = metrics_plot[col] / max_val
                else:
                    # Lower is better, invert and scale
                    max_val = metrics_plot[col].max()
                    if max_val != 0:  # Avoid division by zero
                        metrics_plot[col] = 1 - (metrics_plot[col] / max_val)
            
            # Plot bar chart
            metrics_plot = metrics_plot.reset_index()
            metrics_plot = pd.melt(metrics_plot, id_vars='index', var_name='Metric', value_name='Score')
            metrics_plot.rename(columns={'index': 'Model'}, inplace=True)
            
            plt.figure(figsize=(12, 8))
            sns.barplot(x='Model', y='Score', hue='Metric', data=metrics_plot)
            plt.title('Model Comparison (Higher is Better)')
            plt.ylim(0, 1)
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()
        except Exception as e:
            print(f"Warning: Could not create metrics comparison chart: {str(e)}")
    
    except Exception as e:
        print(f"Warning: Could not create visualization: {str(e)}")
        print("Continuing without visualization...")

###############################################################################
# MAIN EXECUTION PIPELINE
###############################################################################

def run_complete_forecasting_pipeline(data_file):
    """Complete time series forecasting pipeline with all fixes"""
    try:
        # Step 1: Load and preprocess data
        print("\n1. Loading data...")
        data = load_data(data_file)
        print(f"Loaded {len(data)} rows of data")
        
        # Step 2: Analyze the time series
        print("\n2. Analyzing time series...")
        ts_analysis = analyze_time_series(data['Receipt_Count'])
        
        # Step 3: Enhance features
        print("\n3. Enhancing features...")
        enhanced_data = enhance_features(data.copy())
        print(f"Created {len(enhanced_data.columns)} features")
        
        # Step 4: Create proper time series CV splits
        print("\n4. Creating cross-validation splits...")
        cv_splits, (final_train_idx, final_test_idx) = time_series_cv_split(
            enhanced_data, n_splits=5, test_size=0.2
        )
        print(f"Created {len(cv_splits)} CV folds plus a final evaluation split")
        print(f"Final train size: {len(final_train_idx)}, test size: {len(final_test_idx)}")
        
        # Step 5: Train and evaluate models
        print("\n5. Training and evaluating models...")
        
        # 5.1 Baseline models
        print("\n5.1 Creating baseline forecasts...")
        naive_forecast = enhanced_data['Receipt_Count'].iloc[final_train_idx[-1]] * np.ones(len(final_test_idx))
        
        seasonal_naive_forecast = np.array([])
        period = ts_analysis['seasonal_period']
        for i in range(len(final_test_idx)):
            season_idx = (final_train_idx[-1] - period + i % period) % len(enhanced_data)
            seasonal_naive_forecast = np.append(seasonal_naive_forecast, 
                                            enhanced_data['Receipt_Count'].iloc[season_idx])
        
        # 5.2 SARIMA model
        print("\n5.2 Training SARIMA model...")
        try:
            sarima_model = fit_improved_sarima(
                enhanced_data['Receipt_Count'].iloc[final_train_idx], ts_analysis
            )
            sarima_forecast = sarima_model.forecast(steps=len(final_test_idx))
        except Exception as e:
            print(f"Error in SARIMA modeling: {str(e)}")
            # Fallback to naive forecast if SARIMA fails
            sarima_model = None
            sarima_forecast = naive_forecast.copy()
        
        # 5.3 LSTM model
        print("\n5.3 Training LSTM model...")
        try:
            # Prepare scaled data for LSTM
            scaler = MinMaxScaler(feature_range=(0, 1))
            
            # Ensure we're working with numpy arrays
            lstm_data = enhanced_data['Receipt_Count'].iloc[final_train_idx].values.reshape(-1, 1)
            lstm_data_scaled = scaler.fit_transform(lstm_data)
            
            # Create sequences
            X_lstm, y_lstm = [], []
            seq_length = 12
            
            for i in range(len(lstm_data_scaled) - seq_length):
                X_lstm.append(lstm_data_scaled[i:i+seq_length])
                y_lstm.append(lstm_data_scaled[i+seq_length])
            
            X_lstm, y_lstm = np.array(X_lstm), np.array(y_lstm)
            
            # Split for training and validation
            train_size = int(0.8 * len(X_lstm))
            X_train, X_val = X_lstm[:train_size], X_lstm[train_size:]
            y_train, y_val = y_lstm[:train_size], y_lstm[train_size:]
            
            # Reshape for LSTM [samples, time steps, features]
            X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
            X_val = X_val.reshape(X_val.shape[0], X_val.shape[1], 1)
            
            # Create and train LSTM model
            lstm_model, history = create_fixed_lstm_model(X_train, y_train, X_val, y_val)
            
            # Generate LSTM forecasts with improved approach
            lstm_forecast = []
            
            # Get last sequence from training data
            last_sequence = lstm_data[-seq_length:].reshape(-1, 1)
            last_sequence_scaled = scaler.transform(last_sequence)
            curr_sequence = last_sequence_scaled.reshape(1, seq_length, 1)
            
            # Direct forecasting - predict each step but don't use predictions as inputs
            for i in range(len(final_test_idx)):
                # Make one prediction
                next_pred = lstm_model.predict(curr_sequence, verbose=0)[0][0]
                lstm_forecast.append(next_pred)
                
                # Update sequence for next prediction with proper sliding window
                if i < seq_length - 1:
                    # Still have some ground truth data to use
                    last_sequence_scaled = np.roll(last_sequence_scaled, -1, axis=0)
                    # Use ground truth for the newest element if available
                    if i < len(lstm_data) - seq_length:
                        last_sequence_scaled[-1] = lstm_data_scaled[seq_length + i]
                    else:
                        # Otherwise use our prediction
                        last_sequence_scaled[-1] = next_pred
                    curr_sequence = last_sequence_scaled.reshape(1, seq_length, 1)
                else:
                    # Use fully predicted sequence with sliding window
                    curr_sequence = np.roll(curr_sequence, -1, axis=1)
                    curr_sequence[0, -1, 0] = next_pred
            
            # Inverse transform to get original scale
            lstm_forecast = scaler.inverse_transform(
                np.array(lstm_forecast).reshape(-1, 1)
            ).flatten()
        except Exception as e:
            print(f"Error in LSTM modeling: {str(e)}")
            # Fallback to naive forecast if LSTM fails
            lstm_model = None
            lstm_forecast = naive_forecast.copy()
        
        # 5.4 XGBoost model
        print("\n5.4 Training XGBoost model...")
        try:
            xgb_models = improved_xgboost(enhanced_data, ts_analysis, cv_splits)
            
            # Prepare test features
            X_test_xgb = enhanced_data.iloc[final_test_idx].copy()
            feature_cols = [col for col in enhanced_data.columns if col != 'Receipt_Count']
            
            # Fix lag features to avoid data leakage
            for lag in [1, 7, 14, 28]:
                lag_col = f'lag_{lag}'
                if lag_col in X_test_xgb.columns:
                    for i, idx in enumerate(final_test_idx):
                        if idx - lag >= 0:
                            X_test_xgb.iloc[i, X_test_xgb.columns.get_loc(lag_col)] = enhanced_data.iloc[idx - lag]['Receipt_Count']
            
            # Generate predictions from each model
            xgb_forecasts = []
            for model in xgb_models:
                try:
                    # Get the features this model was trained with
                    model_features = [f for f in feature_cols if f in X_test_xgb.columns]
                    xgb_forecasts.append(model.predict(X_test_xgb[model_features]))
                except Exception as e:
                    print(f"Warning: Error predicting with XGBoost model: {str(e)}")
            
            # Use the mean of all model predictions if we have any
            if xgb_forecasts:
                xgb_forecast = np.mean(xgb_forecasts, axis=0)
            else:
                xgb_forecast = naive_forecast.copy()
        except Exception as e:
            print(f"Error in XGBoost modeling: {str(e)}")
            # Fallback to naive forecast if XGBoost fails
            xgb_models = []
            xgb_forecast = naive_forecast.copy()
        
        # Step 6: Evaluate all models
        print("\n6. Evaluating all models...")
        y_true = enhanced_data['Receipt_Count'].iloc[final_test_idx]
        
        forecasts = [naive_forecast, seasonal_naive_forecast, sarima_forecast, lstm_forecast, xgb_forecast]
        model_names = ['Naive', 'Seasonal Naive', 'SARIMA', 'LSTM', 'XGBoost']
        
        # Ensure all forecasts are the same length as y_true
        for i in range(len(forecasts)):
            if len(forecasts[i]) != len(y_true):
                print(f"Warning: {model_names[i]} forecast length mismatch. Adjusting...")
                if len(forecasts[i]) > len(y_true):
                    forecasts[i] = forecasts[i][:len(y_true)]
                else:
                    # Pad with last value if too short
                    forecasts[i] = np.append(forecasts[i], 
                                           np.full(len(y_true) - len(forecasts[i]), forecasts[i][-1]))
        
        evaluation = evaluate_forecasts(y_true, forecasts, model_names)
        
        # Print detailed evaluation
        print("\nModel Performance Comparison:")
        print(evaluation)
        
        # Step 7: Visualize results
        print("\n7. Visualizing results...")
        visualize_forecasts(enhanced_data, final_train_idx, final_test_idx, 
                         forecasts, model_names, evaluation)
        
        return {
            'evaluation': evaluation,
            'models': {
                'sarima': sarima_model,
                'lstm': lstm_model,
                'xgboost': xgb_models
            },
            'forecasts': dict(zip(model_names, forecasts)),
            'data': enhanced_data,
            'train_idx': final_train_idx,
            'test_idx': final_test_idx
        }
    
    except Exception as e:
        print(f"Error in forecasting pipeline: {str(e)}")
        print("Stack trace:")
        import traceback
        traceback.print_exc()
        return None

###############################################################################
# EXECUTE IF RUN AS SCRIPT
###############################################################################

if __name__ == "__main__":
    # Path to the data file
    data_file = "https://fetch-hiring.s3.amazonaws.com/machine-learning-engineer/receipt-count-prediction/data_daily.csv"
    
    # Run the complete pipeline
    results = run_complete_forecasting_pipeline(data_file)
    
    # The results dictionary contains everything you need to further analyze or use the models
    print("\nForecasting pipeline completed successfully!")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss, acf, pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import xgboost as xgb
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from scipy.signal import find_peaks
import warnings
from pathlib import Path

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')
tf.keras.utils.set_random_seed(42)  # For reproducibility

###############################################################################
# DATA LOADING AND PREPROCESSING
###############################################################################

def load_data(file_path):
    """Load time series data from a CSV file"""
    df = pd.read_csv(file_path, parse_dates=['# Date'])
    
    # Rename date column
    df = df.rename(columns={'# Date': 'Date'})
    
    # Set the date column as the index
    df = df.set_index('Date')
    
    # Ensure the data is sorted by date
    df = df.sort_index()
    
    # Do NOT convert to period index since it causes plotting issues
    # Instead ensure it's a DatetimeIndex
    if not isinstance(df.index, pd.DatetimeIndex):
        df.index = pd.to_datetime(df.index)
    
    return df

###############################################################################
# FEATURE ENGINEERING
###############################################################################

def enhance_features(df):
    """Enhanced feature engineering for time series data"""
    # Create a copy to avoid modifying the original
    df = df.copy()
    
    # Ensure proper datetime index
    if isinstance(df.index, pd.PeriodIndex):
        df.index = df.index.to_timestamp()
    elif not isinstance(df.index, pd.DatetimeIndex):
        try:
            df.index = pd.to_datetime(df.index)
        except:
            print("Warning: Could not convert index to DatetimeIndex")
    
    # Calendar features
    try:
        df['dayofweek'] = df.index.dayofweek
        df['day'] = df.index.day  # Day of month
        df['month'] = df.index.month
        df['quarter'] = df.index.quarter
        df['year'] = df.index.year
        df['dayofyear'] = df.index.dayofyear
        
        # Handle potential errors with isocalendar
        try:
            df['weekofyear'] = df.index.isocalendar().week.astype(int)
        except:
            # Fallback if isocalendar attribute error occurs
            df['weekofyear'] = df.index.week
        
        df['is_weekend'] = (df.index.dayofweek >= 5).astype(int)
        df['is_month_end'] = df.index.is_month_end.astype(int)
        df['is_month_start'] = df.index.is_month_start.astype(int)
        df['is_quarter_end'] = df.index.is_quarter_end.astype(int)
        df['is_quarter_start'] = df.index.is_quarter_start.astype(int)
    except Exception as e:
        print(f"Warning: Error creating calendar features: {str(e)}")
    
    # Try to add US holidays if the holidays package is available
    try:
        import holidays
        us_holidays = holidays.US()
        df['is_holiday'] = df.index.to_series().apply(lambda x: x in us_holidays).astype(int)
    except:
        print("Holidays package not available, skipping holiday feature")
    
    # Add day name for easier interpretation
    day_names = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 
                4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
    if 'dayofweek' in df.columns:
        df['dayname'] = df['dayofweek'].map(day_names)
    
    # Rolling statistics - careful about leakage!
    df['rolling_mean_7d'] = df['Receipt_Count'].rolling(window=7, min_periods=1).mean()
    df['rolling_std_7d'] = df['Receipt_Count'].rolling(window=7, min_periods=1).std()
    df['rolling_mean_28d'] = df['Receipt_Count'].rolling(window=28, min_periods=1).mean()
    
    # Lag features - be careful about leakage when using these!
    for lag in [1, 7, 14, 28]:
        df[f'lag_{lag}'] = df['Receipt_Count'].shift(lag)
    
    # Handle missing values from rolling/lag operations
    df = df.fillna(method='bfill')
    
    return df

###############################################################################
# TIME SERIES ANALYSIS
###############################################################################

def analyze_time_series(data, plot=True):
    """Improved analysis for stationarity and seasonality"""
    # Ensure we're working with a Series
    if isinstance(data, pd.DataFrame):
        data = data['Receipt_Count']
    
    print("Time Series Analysis:")
    print(f"Data points: {len(data)}")
    print(f"Start date: {data.index[0]}")
    print(f"End date: {data.index[-1]}")
    print(f"Min value: {data.min():.2f}, Max value: {data.max():.2f}")
    print(f"Mean: {data.mean():.2f}, Median: {data.median():.2f}")
    
    # Test for stationarity with ADF test
    adf_result = adfuller(data.dropna())
    print("\nAugmented Dickey-Fuller Test:")
    print(f"ADF Statistic: {adf_result[0]:.4f}")
    print(f"p-value: {adf_result[1]:.4f}")
    print(f"Critical Values: {adf_result[4]}")
    is_stationary = adf_result[1] < 0.05
    print(f"Is stationary: {is_stationary}")
    
    # Determine seasonality using autocorrelation
    acf_values = acf(data.dropna(), nlags=365, fft=True)
    
    # Find peaks in ACF
    peaks, _ = find_peaks(acf_values)
    
    seasonal_periods = []
    if len(peaks) > 0:
        # Find significant peaks
        for peak in peaks:
            if peak > 1 and acf_values[peak] > 0.2:  # Threshold for significance
                seasonal_periods.append(peak)
                if len(seasonal_periods) >= 3:
                    break
        
        print("\nPotential seasonal periods detected in ACF:")
        for period in seasonal_periods:
            print(f"- Period of {period} days (correlation: {acf_values[period]:.3f})")
    
    # Check for weekly, monthly, and yearly patterns
    seasonal_results = {}
    for period in [7, 30, 90, 365]:  # Weekly, monthly, quarterly, yearly
        try:
            result = seasonal_decompose(data, model='additive', period=period)
            seasonal_variance = np.var(result.seasonal)
            seasonal_results[period] = seasonal_variance
            print(f"Seasonal variance with period={period}: {seasonal_variance:.4f}")
        except:
            print(f"Could not decompose with period={period}")
    
    # Select strongest seasonal period
    best_period = 7  # Default weekly
    if seasonal_results:
        best_period = max(seasonal_results.items(), key=lambda x: x[1])[0]
        print(f"Best seasonal period based on variance: {best_period}")
    
    # Plot the decomposition using the best period
    if plot and seasonal_results:
        plt.figure(figsize=(15, 12))
        result = seasonal_decompose(data, model='additive', period=best_period)
        result.plot()
        plt.tight_layout()
        plt.show()
        
        # Plot autocorrelation function
        plt.figure(figsize=(15, 6))
        plot_acf(data, lags=50, title="Autocorrelation Function")
        plt.grid(True)
        plt.show()
        
        # Plot partial autocorrelation function
        plt.figure(figsize=(15, 6))
        plot_pacf(data, lags=50, title="Partial Autocorrelation Function")
        plt.grid(True)
        plt.show()
    
    return {
        "is_stationary": is_stationary, 
        "seasonal_period": best_period, 
        "detected_periods": seasonal_periods,
        "seasonal_variances": seasonal_results
    }

###############################################################################
# TIME SERIES CROSS-VALIDATION
###############################################################################

def time_series_cv_split(data, n_splits=5, test_size=0.2):
    """Create properly separated time series folds"""
    # Calculate the size of each fold
    fold_size = int(len(data) / (n_splits + 1))  # +1 for final evaluation
    
    cv_splits = []
    for i in range(n_splits):
        # Calculate start and end points for each fold
        train_end = (i + 1) * fold_size
        test_start = train_end
        test_end = test_start + fold_size
        
        # Safety checks for indices
        test_end = min(test_end, len(data))
        
        # Create train/test indices
        train_indices = list(range(0, train_end))
        test_indices = list(range(test_start, test_end))
        
        cv_splits.append((train_indices, test_indices))
    
    # Create final train/test split for evaluation
    train_size = int(len(data) * (1 - test_size))
    final_train = list(range(0, train_size))
    final_test = list(range(train_size, len(data)))
    
    return cv_splits, (final_train, final_test)

###############################################################################
# SEQUENCE CREATION FOR LSTM
###############################################################################

def create_sequences(data, seq_length=12, target_col='Receipt_Count'):
    """Create sequences for LSTM model with multiple features"""
    # Convert DataFrame to numpy arrays
    # Features are all columns except the target
    feature_cols = [col for col in data.columns if col != target_col]
    
    X, y = [], []
    
    for i in range(len(data) - seq_length):
        # For X, use all features
        features_seq = data.iloc[i:i+seq_length][feature_cols].values
        # For y, use only the target column
        target = data.iloc[i+seq_length][target_col]
        
        X.append(features_seq)
        y.append(target)
    
    return np.array(X), np.array(y)

###############################################################################
# MODEL IMPLEMENTATIONS
###############################################################################

def fit_improved_sarima(data, seasonal_analysis):
    """Improved SARIMA with proper parameter selection"""
    # Get seasonal period from analysis
    seasonal_period = seasonal_analysis["seasonal_period"]
    
    print(f"Fitting SARIMA with seasonal period: {seasonal_period}")
    
    # Try to use auto_arima if available, otherwise use fixed parameters
    try:
        from pmdarima import auto_arima
        
        print("Using auto_arima to find best parameters...")
        auto_model = auto_arima(
            data,
            seasonal=True,
            m=seasonal_period,
            start_p=0, max_p=3,
            start_q=0, max_q=3,
            start_P=0, max_P=2,
            start_Q=0, max_Q=2,
            d=None, D=None,  # Allow auto_arima to determine
            trace=True,
            error_action='ignore',
            suppress_warnings=True,
            stepwise=True
        )
        
        order = auto_model.order
        seasonal_order = auto_model.seasonal_order
        print(f"Best SARIMA parameters: {order} {seasonal_order}")
        
    except ImportError:
        # Fallback to standard parameters if auto_arima is not available
        print("auto_arima not available, using default parameters")
        order = (1, 1, 1)
        seasonal_order = (1, 1, 1, seasonal_period)
    
    # Fit final model with best parameters
    print("Fitting final SARIMA model...")
    final_model = sm.tsa.statespace.SARIMAX(
        data,
        order=order,
        seasonal_order=seasonal_order,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    
    final_model_fit = final_model.fit(disp=False)
    print(f"SARIMA model AIC: {final_model_fit.aic:.2f}")
    
    return final_model_fit

def prepare_multivariate_lstm_data(data, seq_length=30, pred_length=1):
    """Create multivariate sequences for LSTM with more features"""
    print("Preparing multivariate LSTM data...")
    
    # Select relevant columns for the LSTM model
    # Target column and important predictors
    selected_cols = [
        'Receipt_Count',  # Target variable
        'dayofweek', 'month', 'is_weekend',  # Calendar features
        'rolling_mean_7d', 'rolling_std_7d',  # Rolling statistics
    ]
    
    try:
        # Add more columns if they exist
        if 'is_holiday' in data.columns:
            selected_cols.append('is_holiday')
        if 'is_month_end' in data.columns:
            selected_cols.append('is_month_end')
        if 'is_quarter_end' in data.columns:
            selected_cols.append('is_quarter_end')
        
        # Add derived features for LSTM
        data = data.copy()
        # Cyclical encoding of time features
        data['dayofweek_sin'] = np.sin(2 * np.pi * data['dayofweek']/7)
        data['dayofweek_cos'] = np.cos(2 * np.pi * data['dayofweek']/7)
        data['month_sin'] = np.sin(2 * np.pi * data['month']/12)
        data['month_cos'] = np.cos(2 * np.pi * data['month']/12)
        
        selected_cols.extend(['dayofweek_sin', 'dayofweek_cos', 'month_sin', 'month_cos'])
    except Exception as e:
        print(f"Warning: Error adding derived features: {str(e)}")
    
    # Filter and clean the data
    lstm_data = data[selected_cols].copy()
    lstm_data = lstm_data.fillna(method='bfill').fillna(method='ffill')
    
    # Create scalers for each feature
    scalers = {}
    scaled_data = lstm_data.copy()
    
    # Scale each column independently
    for col in lstm_data.columns:
        scaler = MinMaxScaler(feature_range=(0, 1))
        scaled_data[col] = scaler.fit_transform(lstm_data[col].values.reshape(-1, 1)).flatten()
        scalers[col] = scaler
    
    # Create sequences
    X, y = [], []
    target_scaler = scalers['Receipt_Count']
    
    for i in range(len(scaled_data) - seq_length - pred_length + 1):
        # Input sequence (all features)
        X.append(scaled_data.iloc[i:i+seq_length].values)
        # Target is next 'pred_length' values of Receipt_Count
        if pred_length == 1:
            y.append(scaled_data.iloc[i+seq_length]['Receipt_Count'])
        else:
            y.append(scaled_data.iloc[i+seq_length:i+seq_length+pred_length]['Receipt_Count'].values)
    
    X = np.array(X)
    y = np.array(y)
    
    print(f"Created {len(X)} sequences with shape {X.shape}")
    
    return X, y, scalers, lstm_data.columns.tolist()

def create_fixed_lstm_model(X_train, y_train, X_val, y_val, multi_step=False):
    """Significantly improved LSTM model with multivariate inputs and advanced architecture"""
    # Get input dimensions
    n_features = X_train.shape[2]
    seq_length = X_train.shape[1]
    
    print(f"Creating enhanced LSTM model with input shape: (samples, {seq_length}, {n_features})")
    
    # Determine output shape (single value or sequence)
    if multi_step and len(y_train.shape) > 1:
        output_length = y_train.shape[1]
        print(f"Multi-step forecasting with {output_length} steps")
    else:
        output_length = 1
        print("Single-step forecasting")
    
    # Create a more robust and sophisticated model architecture
    model = Sequential()
    
    # First Bidirectional LSTM layer with larger units for better pattern recognition
    model.add(tf.keras.layers.Bidirectional(
        LSTM(128, activation='tanh', recurrent_activation='sigmoid', 
             return_sequences=True, recurrent_dropout=0.1),
        input_shape=(seq_length, n_features)
    ))
    model.add(BatchNormalization())
    model.add(Dropout(0.3))
    
    # Second Bidirectional LSTM layer
    model.add(tf.keras.layers.Bidirectional(
        LSTM(64, activation='tanh', recurrent_activation='sigmoid', 
             return_sequences=False, recurrent_dropout=0.1)
    ))
    model.add(BatchNormalization())
    model.add(Dropout(0.3))
    
    # Dense layers with residual connections
    model.add(Dense(64, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    
    model.add(Dense(32, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    
    # Output layer - either single value or sequence
    if output_length > 1:
        model.add(Dense(output_length))
    else:
        model.add(Dense(1))
    
    # Compile with Huber loss for robustness to outliers
    model.compile(
        loss=tf.keras.losses.Huber(delta=1.0),  # Huber loss is more robust than MSE
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005)  # Lower learning rate
    )
    
    # Show model summary
    model.summary()
    
    # Callbacks for better training
    callbacks = [
        EarlyStopping(patience=20, restore_best_weights=True, monitor='val_loss'),
        ReduceLROnPlateau(factor=0.2, patience=5, min_lr=0.00001, monitor='val_loss'),
        tf.keras.callbacks.ModelCheckpoint('best_lstm_model.h5', save_best_only=True, monitor='val_loss')
    ]
    
    # Train with validation data and class weights
    print("Training LSTM model...")
    history = model.fit(
        X_train, y_train,
        epochs=200,  # More epochs with early stopping
        batch_size=64,  # Larger batch size
        validation_data=(X_val, y_val),
        callbacks=callbacks,
        verbose=1,
        shuffle=True  # Important for time series
    )
    
    # Plot training history
    try:
        plt.figure(figsize=(12, 6))
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('LSTM Model Training History')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True)
        plt.yscale('log')  # Log scale for better visualization of convergence
        plt.show()
    except Exception as e:
        print(f"Warning: Could not plot training history: {str(e)}")
    
    # Try to load the best model if it was saved
    try:
        best_model = tf.keras.models.load_model('best_lstm_model.h5')
        print("Loaded best model from checkpoint")
        return best_model, history
    except:
        print("Using final model (checkpoint loading failed)")
        return model, history

def lstm_ensemble_forecast(models, scalers, feature_columns, history_data, forecast_steps, seq_length=30):
    """Ensemble forecasting approach for LSTM with multiple models and multivariate inputs"""
    print(f"Generating {forecast_steps} step forecast with LSTM ensemble...")
    
    # Create a copy of the history data for predictions
    forecast_data = history_data.copy()
    
    # Ensure we have at least seq_length data points
    if len(forecast_data) < seq_length:
        raise ValueError(f"History data must contain at least {seq_length} points")
    
    # Create a container for all model predictions
    all_model_forecasts = []
    
    # For each model in the ensemble
    for i, (model, scaler) in enumerate(zip(models, scalers)):
        print(f"  Generating forecasts with model {i+1}...")
        
        # Create a working copy of the data
        model_forecast_data = forecast_data.copy()
        
        # Scale the data
        scaled_data = model_forecast_data.copy()
        for col in feature_columns:
            if col in scaled_data.columns:
                scaled_data[col] = scaler.transform(
                    scaled_data[col].values.reshape(-1, 1)
                ).flatten()
        
        # Generate forecasts for each timestep
        model_forecasts = []
        
        for step in range(forecast_steps):
            # Get the last seq_length values as input sequence
            input_seq = scaled_data.iloc[-seq_length:][feature_columns].values
            input_seq = input_seq.reshape(1, seq_length, len(feature_columns))
            
            # Make prediction
            pred = model.predict(input_seq, verbose=0)[0]
            # If prediction is multi-step, take only the first value
            if hasattr(pred, "__len__") and len(pred) > 1:
                pred = pred[0]
                
            model_forecasts.append(pred)
            
            # Update data for next step prediction
            # Create a new row with the predicted value
            new_row = model_forecast_data.iloc[-1:].copy()
            new_row.index = [model_forecast_data.index[-1] + pd.Timedelta(days=1)]
            
            # Set the Receipt_Count to the predicted value (inverse transform)
            receipt_col_idx = feature_columns.index('Receipt_Count')
            pred_array = np.zeros((1, len(feature_columns)))
            pred_array[0, receipt_col_idx] = pred
            
            # Get the original scale value for Receipt_Count
            inverse_pred = scaler.inverse_transform(pred_array)[0, receipt_col_idx]
            new_row['Receipt_Count'] = inverse_pred
            
            # Update other time features for the new row
            if 'dayofweek' in new_row.columns:
                new_row['dayofweek'] = new_row.index[0].dayofweek
            if 'month' in new_row.columns:
                new_row['month'] = new_row.index[0].month
            if 'is_weekend' in new_row.columns:
                new_row['is_weekend'] = (new_row.index[0].dayofweek >= 5).astype(int)
            if 'day' in new_row.columns:
                new_row['day'] = new_row.index[0].day
            if 'dayofyear' in new_row.columns:
                new_row['dayofyear'] = new_row.index[0].dayofyear
            
            # Update rolling statistics if available
            if 'rolling_mean_7d' in new_row.columns:
                new_values = pd.concat([model_forecast_data.iloc[-6:], new_row])
                new_row['rolling_mean_7d'] = new_values['Receipt_Count'].mean()
            
            if 'rolling_std_7d' in new_row.columns:
                new_values = pd.concat([model_forecast_data.iloc[-6:], new_row])
                new_row['rolling_std_7d'] = new_values['Receipt_Count'].std()
            
            # Add the new row to the forecast data
            model_forecast_data = pd.concat([model_forecast_data, new_row])
            
            # Update the scaled data for next prediction
            scaled_data = model_forecast_data.copy()
            for col in feature_columns:
                if col in scaled_data.columns:
                    scaled_data[col] = scaler.transform(
                        scaled_data[col].values.reshape(-1, 1)
                    ).flatten()
        
        # Add this model's forecasts to the ensemble
        all_model_forecasts.append(model_forecasts)
    
    # If we have multiple models, average their predictions
    if len(all_model_forecasts) > 1:
        ensemble_forecast = np.mean(all_model_forecasts, axis=0)
    else:
        ensemble_forecast = all_model_forecasts[0]
    
    # Convert predictions back to original scale if they aren't already
    if 'Receipt_Count' in scalers:
        ensemble_forecast = scalers['Receipt_Count'].inverse_transform(
            np.array(ensemble_forecast).reshape(-1, 1)
        ).flatten()
    
    return ensemble_forecast

def improved_xgboost(data, ts_analysis, cv_splits):
    """XGBoost with improved features and hyperparameter tuning"""
    print("Training XGBoost model with improved features...")
    
    # Create advanced features specifically for XGBoost
    data = data.copy()
    
    # Add exponential weighted moving averages
    data['ewma_7d'] = data['Receipt_Count'].ewm(span=7).mean()
    data['ewma_14d'] = data['Receipt_Count'].ewm(span=14).mean()
    data['ewma_30d'] = data['Receipt_Count'].ewm(span=30).mean()
    
    # Add rolling statistics
    for window in [7, 14, 30, 90]:
        data[f'roll_mean_{window}d'] = data['Receipt_Count'].rolling(window=window, min_periods=1).mean()
        data[f'roll_std_{window}d'] = data['Receipt_Count'].rolling(window=window, min_periods=1).std()
        data[f'roll_max_{window}d'] = data['Receipt_Count'].rolling(window=window, min_periods=1).max()
        data[f'roll_min_{window}d'] = data['Receipt_Count'].rolling(window=window, min_periods=1).min()
    
    # Add trend indicators
    data['trend_7d'] = data['Receipt_Count'].diff(7)
    data['trend_14d'] = data['Receipt_Count'].diff(14)
    data['trend_30d'] = data['Receipt_Count'].diff(30)
    
    # Add day of year cyclical features using sine and cosine transforms
    data['dayofyear_sin'] = np.sin(2 * np.pi * data.index.dayofyear/365.25)
    data['dayofyear_cos'] = np.cos(2 * np.pi * data.index.dayofyear/365.25)
    
    # Add month cyclical features
    data['month_sin'] = np.sin(2 * np.pi * data.index.month/12)
    data['month_cos'] = np.cos(2 * np.pi * data.index.month/12)
    
    # Add day of week cyclical features
    data['dayofweek_sin'] = np.sin(2 * np.pi * data.index.dayofweek/7)
    data['dayofweek_cos'] = np.cos(2 * np.pi * data.index.dayofweek/7)
    
    # Add ratio features
    data['ratio_7_30'] = data['roll_mean_7d'] / data['roll_mean_30d']
    data['ratio_7_90'] = data['roll_mean_7d'] / data['roll_mean_90d']
    
    # Fill NaN values
    data = data.fillna(method='bfill').fillna(method='ffill')
    
    # Prepare feature columns (exclude the target)
    feature_cols = [col for col in data.columns if col != 'Receipt_Count']
    
    # Create feature sets for each fold to prevent leakage
    fold_models = []
    fold_metrics = []
    
    # Store feature importances
    all_importances = {}
    
    for fold_idx, (train_idx, test_idx) in enumerate(cv_splits):
        print(f"\nTraining XGBoost model for fold {fold_idx+1}/{len(cv_splits)}...")
        
        # Split data
        train_data = data.iloc[train_idx]
        test_data = data.iloc[test_idx]
        
        # Define X and y
        X_train = train_data[feature_cols]
        y_train = train_data['Receipt_Count']
        X_test = test_data[feature_cols]
        y_test = test_data['Receipt_Count']
        
        # Train XGBoost model with optimized parameters
        model = xgb.XGBRegressor(
            objective='reg:squarederror',
            max_depth=6,              # Increased depth for more complex patterns
            learning_rate=0.05,       # Lower learning rate for better convergence
            n_estimators=500,         # More estimators for better performance
            subsample=0.8,
            colsample_bytree=0.8,
            min_child_weight=3,       # Controls model complexity
            gamma=0.1,                # Minimum loss reduction to create a new tree split
            reg_alpha=0.1,            # L1 regularization
            reg_lambda=1.0,           # L2 regularization
            random_state=42
        )
        
        # Fit model - handle compatibility issues with different XGBoost versions
        try:
            # Try with early_stopping_rounds parameter
            eval_set = [(X_test, y_test)]
            model.fit(X_train, y_train, eval_set=eval_set, early_stopping_rounds=30, verbose=False)
        except TypeError:
            try:
                # Try with eval_metric parameter
                model.fit(X_train, y_train, eval_set=eval_set, eval_metric='rmse', verbose=False)
            except:
                # Fallback to basic fit
                model.fit(X_train, y_train)
        
        # Evaluate
        y_pred = model.predict(X_test)
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        mae = mean_absolute_error(y_test, y_pred)
        
        print(f"  Fold {fold_idx+1} - RMSE: {rmse:.2f}, MAE: {mae:.2f}")
        
        # Add to fold results
        fold_models.append(model)
        fold_metrics.append(rmse)
        
        # Store feature importances
        for feature, importance in zip(feature_cols, model.feature_importances_):
            if feature not in all_importances:
                all_importances[feature] = []
            all_importances[feature].append(importance)
    
    # Report average performance
    avg_rmse = np.mean(fold_metrics)
    std_rmse = np.std(fold_metrics)
    print(f"XGBoost CV RMSE: {avg_rmse:.2f} ± {std_rmse:.2f}")
    
    # Create a feature importance summary if we have results
    if all_importances:
        feature_importance = {feature: np.mean(importances) for feature, importances in all_importances.items()}
        importance_df = pd.DataFrame({
            'Feature': list(feature_importance.keys()),
            'Importance': list(feature_importance.values())
        }).sort_values('Importance', ascending=False)
        
        try:
            plt.figure(figsize=(12, 8))
            sns.barplot(x='Importance', y='Feature', data=importance_df.head(15))
            plt.title('XGBoost Feature Importance (Top 15)')
            plt.tight_layout()
            plt.show()
        except Exception as e:
            print(f"Couldn't plot feature importance: {str(e)}")
    
    # Return the ensemble of models
    return fold_models

###############################################################################
# EVALUATION FUNCTIONS
###############################################################################

def evaluate_forecasts(y_true, forecasts, model_names):
    """Comprehensive evaluation with multiple metrics"""
    metrics = {}
    
    for name, forecast in zip(model_names, forecasts):
        # Standard metrics
        rmse = np.sqrt(mean_squared_error(y_true, forecast))
        mae = mean_absolute_error(y_true, forecast)
        r2 = r2_score(y_true, forecast)
        
        # Direction accuracy (for trends)
        direction_true = np.diff(y_true) > 0
        direction_pred = np.diff(forecast) > 0
        direction_accuracy = np.mean(direction_true == direction_pred)
        
        # MAPE (handle zeros carefully)
        mask = y_true != 0
        if any(mask):
            mape = np.mean(np.abs((y_true[mask] - forecast[mask]) / y_true[mask])) * 100
        else:
            mape = np.nan
        
        # Store metrics
        metrics[name] = {
            'RMSE': rmse,
            'MAE': mae,
            'R²': r2,
            'Direction Accuracy': direction_accuracy,
            'MAPE': mape
        }
    
    # Create comparison dataframe
    metrics_df = pd.DataFrame(metrics).T
    
    # Calculate ranks
    for col in metrics_df.columns:
        # Lower is better for error metrics, higher better for R² and direction
        if col in ['R²', 'Direction Accuracy']:
            metrics_df[f'{col} Rank'] = metrics_df[col].rank(ascending=False)
        else:
            metrics_df[f'{col} Rank'] = metrics_df[col].rank()
    
    # Add average rank
    rank_cols = [col for col in metrics_df.columns if 'Rank' in col]
    metrics_df['Avg Rank'] = metrics_df[rank_cols].mean(axis=1)
    
    # Sort by average rank
    metrics_df = metrics_df.sort_values('Avg Rank')
    
    return metrics_df

###############################################################################
# VISUALIZATION FUNCTIONS
###############################################################################

def visualize_forecasts(data, train_idx, test_idx, forecasts, model_names, metrics_df):
    """Visualize time series forecasts with comprehensive plots"""
    try:
        # Extract train and test data
        train_data = data.iloc[train_idx]['Receipt_Count']
        test_data = data.iloc[test_idx]['Receipt_Count']
        
        # Ensure datetime index
        if isinstance(train_data.index, pd.PeriodIndex):
            train_data.index = train_data.index.to_timestamp()
        if isinstance(test_data.index, pd.PeriodIndex):
            test_data.index = test_data.index.to_timestamp()
        
        # Create figure
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12), 
                                      gridspec_kw={'height_ratios': [3, 1]})
        
        # Plot original data
        ax1.plot(train_data.index, train_data.values, label='Training Data', color='blue')
        ax1.plot(test_data.index, test_data.values, label='Test Data', color='green')
        
        # Define colors for each model
        colors = {
            'Naive': 'gray',
            'Seasonal Naive': 'lightgray',
            'SARIMA': 'purple',
            'LSTM': 'orange',
            'XGBoost': 'red'
        }
        
        # Store prediction errors for the error plot
        prediction_errors = {}
        
        # Plot forecasts
        for name, forecast in zip(model_names, forecasts):
            color = colors.get(name, 'black')
            ax1.plot(test_data.index, forecast, 
                    label=f'{name} Forecast', 
                    color=color,
                    linestyle='--', 
                    linewidth=2)
            
            # Calculate errors for the error plot
            errors = test_data.values - forecast
            prediction_errors[name] = errors
        
        # Plot prediction errors in the second subplot
        for name, errors in prediction_errors.items():
            color = colors.get(name, 'black')
            ax2.plot(test_data.index, errors, 
                    label=f'{name} Error', 
                    color=color,
                    alpha=0.7)
        
        # Add vertical line for train-test split
        split_date = train_data.index[-1]
        ax1.axvline(x=split_date, color='black', linestyle='--')
        ax2.axvline(x=split_date, color='black', linestyle='--')
        
        # Add horizontal line at zero for the error plot
        ax2.axhline(y=0, color='black', linestyle='-', alpha=0.3)
        
        # Format the x-axis to show dates nicely
        for ax in [ax1, ax2]:
            # Simplified date formatting to avoid conversion errors
            ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
            ax.xaxis.set_major_locator(mdates.MonthLocator())
            plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)
            ax.grid(True, alpha=0.3)
        
        # Add titles and labels
        ax1.set_title('Time Series Model Comparison', fontsize=16)
        ax1.set_ylabel('Receipt Count', fontsize=12)
        ax1.legend(loc='upper left', frameon=True, framealpha=0.8, fontsize=10)
        
        ax2.set_title('Prediction Errors', fontsize=14)
        ax2.set_xlabel('Date', fontsize=12)
        ax2.set_ylabel('Error', fontsize=12)
        ax2.legend(loc='upper left', frameon=True, framealpha=0.8, fontsize=10)
        
        # Add a text box with model metrics in the main plot
        metrics_text = "Test Set Metrics:\n"
        for name in model_names:
            rmse = metrics_df.loc[name, 'RMSE']
            mae = metrics_df.loc[name, 'MAE']
            metrics_text += f"{name}: RMSE={rmse:.0f}, MAE={mae:.0f}\n"
        
        # Position the text box in the upper right of the plot
        props = dict(boxstyle='round', facecolor='white', alpha=0.8)
        ax1.text(0.02, 0.02, metrics_text, transform=ax1.transAxes, fontsize=10,
                verticalalignment='bottom', horizontalalignment='left', bbox=props)
        
        plt.tight_layout()
        plt.show()
        
        # Also plot a metrics comparison chart
        try:
            plt.figure(figsize=(12, 8))
            metrics_plot = metrics_df[['RMSE', 'MAE', 'R²', 'Direction Accuracy']].copy()
            
            # Normalize for plotting
            for col in metrics_plot.columns:
                if col in ['R²', 'Direction Accuracy']:
                    # Higher is better, scale to 0-1
                    max_val = metrics_plot[col].max()
                    if max_val != 0:  # Avoid division by zero
                        metrics_plot[col] = metrics_plot[col] / max_val
                else:
                    # Lower is better, invert and scale
                    max_val = metrics_plot[col].max()
                    if max_val != 0:  # Avoid division by zero
                        metrics_plot[col] = 1 - (metrics_plot[col] / max_val)
            
            # Plot bar chart
            metrics_plot = metrics_plot.reset_index()
            metrics_plot = pd.melt(metrics_plot, id_vars='index', var_name='Metric', value_name='Score')
            metrics_plot.rename(columns={'index': 'Model'}, inplace=True)
            
            plt.figure(figsize=(12, 8))
            sns.barplot(x='Model', y='Score', hue='Metric', data=metrics_plot)
            plt.title('Model Comparison (Higher is Better)')
            plt.ylim(0, 1)
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()
        except Exception as e:
            print(f"Warning: Could not create metrics comparison chart: {str(e)}")
    
    except Exception as e:
        print(f"Warning: Could not create visualization: {str(e)}")
        print("Continuing without visualization...")

###############################################################################
# MAIN EXECUTION PIPELINE
###############################################################################

def run_complete_forecasting_pipeline(data_file):
    """Complete time series forecasting pipeline with all fixes"""
    try:
        # Step 1: Load and preprocess data
        print("\n1. Loading data...")
        data = load_data(data_file)
        print(f"Loaded {len(data)} rows of data")
        
        # Step 2: Analyze the time series
        print("\n2. Analyzing time series...")
        ts_analysis = analyze_time_series(data['Receipt_Count'])
        
        # Step 3: Enhance features
        print("\n3. Enhancing features...")
        enhanced_data = enhance_features(data.copy())
        print(f"Created {len(enhanced_data.columns)} features")
        
        # Step 4: Create proper time series CV splits
        print("\n4. Creating cross-validation splits...")
        cv_splits, (final_train_idx, final_test_idx) = time_series_cv_split(
            enhanced_data, n_splits=5, test_size=0.2
        )
        print(f"Created {len(cv_splits)} CV folds plus a final evaluation split")
        print(f"Final train size: {len(final_train_idx)}, test size: {len(final_test_idx)}")
        
        # Step 5: Train and evaluate models
        print("\n5. Training and evaluating models...")
        
        # 5.1 Baseline models
        print("\n5.1 Creating baseline forecasts...")
        naive_forecast = enhanced_data['Receipt_Count'].iloc[final_train_idx[-1]] * np.ones(len(final_test_idx))
        
        seasonal_naive_forecast = np.array([])
        period = ts_analysis['seasonal_period']
        for i in range(len(final_test_idx)):
            season_idx = (final_train_idx[-1] - period + i % period) % len(enhanced_data)
            seasonal_naive_forecast = np.append(seasonal_naive_forecast, 
                                            enhanced_data['Receipt_Count'].iloc[season_idx])
        
        # 5.2 SARIMA model
        print("\n5.2 Training SARIMA model...")
        try:
            sarima_model = fit_improved_sarima(
                enhanced_data['Receipt_Count'].iloc[final_train_idx], ts_analysis
            )
            sarima_forecast = sarima_model.forecast(steps=len(final_test_idx))
        except Exception as e:
            print(f"Error in SARIMA modeling: {str(e)}")
            # Fallback to naive forecast if SARIMA fails
            sarima_model = None
            sarima_forecast = naive_forecast.copy()
        
        # 5.3 LSTM model
        print("\n5.3 Training LSTM model...")
        try:
            # Prepare scaled data for LSTM
            scaler = MinMaxScaler(feature_range=(0, 1))
            
            # Ensure we're working with numpy arrays
            lstm_data = enhanced_data['Receipt_Count'].iloc[final_train_idx].values.reshape(-1, 1)
            lstm_data_scaled = scaler.fit_transform(lstm_data)
            
            # Create sequences
            X_lstm, y_lstm = [], []
            seq_length = 12
            
            for i in range(len(lstm_data_scaled) - seq_length):
                X_lstm.append(lstm_data_scaled[i:i+seq_length])
                y_lstm.append(lstm_data_scaled[i+seq_length])
            
            X_lstm, y_lstm = np.array(X_lstm), np.array(y_lstm)
            
            # Split for training and validation
            train_size = int(0.8 * len(X_lstm))
            X_train, X_val = X_lstm[:train_size], X_lstm[train_size:]
            y_train, y_val = y_lstm[:train_size], y_lstm[train_size:]
            
            # Reshape for LSTM [samples, time steps, features]
            X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
            X_val = X_val.reshape(X_val.shape[0], X_val.shape[1], 1)
            
            # Create and train LSTM model
            lstm_model, history = create_fixed_lstm_model(X_train, y_train, X_val, y_val)
            
            # Generate LSTM forecasts with improved approach
            lstm_forecast = []
            
            # Get last sequence from training data
            last_sequence = lstm_data[-seq_length:].reshape(-1, 1)
            last_sequence_scaled = scaler.transform(last_sequence)
            curr_sequence = last_sequence_scaled.reshape(1, seq_length, 1)
            
            # Direct forecasting - predict each step but don't use predictions as inputs
            for i in range(len(final_test_idx)):
                # Make one prediction
                next_pred = lstm_model.predict(curr_sequence, verbose=0)[0][0]
                lstm_forecast.append(next_pred)
                
                # Update sequence for next prediction with proper sliding window
                if i < seq_length - 1:
                    # Still have some ground truth data to use
                    last_sequence_scaled = np.roll(last_sequence_scaled, -1, axis=0)
                    # Use ground truth for the newest element if available
                    if i < len(lstm_data) - seq_length:
                        last_sequence_scaled[-1] = lstm_data_scaled[seq_length + i]
                    else:
                        # Otherwise use our prediction
                        last_sequence_scaled[-1] = next_pred
                    curr_sequence = last_sequence_scaled.reshape(1, seq_length, 1)
                else:
                    # Use fully predicted sequence with sliding window
                    curr_sequence = np.roll(curr_sequence, -1, axis=1)
                    curr_sequence[0, -1, 0] = next_pred
            
            # Inverse transform to get original scale
            lstm_forecast = scaler.inverse_transform(
                np.array(lstm_forecast).reshape(-1, 1)
            ).flatten()
        except Exception as e:
            print(f"Error in LSTM modeling: {str(e)}")
            # Fallback to naive forecast if LSTM fails
            lstm_model = None
            lstm_forecast = naive_forecast.copy()
        
        # 5.4 XGBoost model
        print("\n5.4 Training XGBoost model...")
        try:
            xgb_models = improved_xgboost(enhanced_data, ts_analysis, cv_splits)
            
            # Prepare test features
            X_test_xgb = enhanced_data.iloc[final_test_idx].copy()
            feature_cols = [col for col in enhanced_data.columns if col != 'Receipt_Count']
            
            # Fix lag features to avoid data leakage
            for lag in [1, 7, 14, 28]:
                lag_col = f'lag_{lag}'
                if lag_col in X_test_xgb.columns:
                    for i, idx in enumerate(final_test_idx):
                        if idx - lag >= 0:
                            X_test_xgb.iloc[i, X_test_xgb.columns.get_loc(lag_col)] = enhanced_data.iloc[idx - lag]['Receipt_Count']
            
            # Generate predictions from each model
            xgb_forecasts = []
            for model in xgb_models:
                try:
                    # Get the features this model was trained with
                    model_features = [f for f in feature_cols if f in X_test_xgb.columns]
                    xgb_forecasts.append(model.predict(X_test_xgb[model_features]))
                except Exception as e:
                    print(f"Warning: Error predicting with XGBoost model: {str(e)}")
            
            # Use the mean of all model predictions if we have any
            if xgb_forecasts:
                xgb_forecast = np.mean(xgb_forecasts, axis=0)
            else:
                xgb_forecast = naive_forecast.copy()
        except Exception as e:
            print(f"Error in XGBoost modeling: {str(e)}")
            # Fallback to naive forecast if XGBoost fails
            xgb_models = []
            xgb_forecast = naive_forecast.copy()
        
        # Step 6: Evaluate all models
        print("\n6. Evaluating all models...")
        y_true = enhanced_data['Receipt_Count'].iloc[final_test_idx]
        
        forecasts = [naive_forecast, seasonal_naive_forecast, sarima_forecast, lstm_forecast, xgb_forecast]
        model_names = ['Naive', 'Seasonal Naive', 'SARIMA', 'LSTM', 'XGBoost']
        
        # Ensure all forecasts are the same length as y_true
        for i in range(len(forecasts)):
            if len(forecasts[i]) != len(y_true):
                print(f"Warning: {model_names[i]} forecast length mismatch. Adjusting...")
                if len(forecasts[i]) > len(y_true):
                    forecasts[i] = forecasts[i][:len(y_true)]
                else:
                    # Pad with last value if too short
                    forecasts[i] = np.append(forecasts[i], 
                                           np.full(len(y_true) - len(forecasts[i]), forecasts[i][-1]))
        
        evaluation = evaluate_forecasts(y_true, forecasts, model_names)
        
        # Print detailed evaluation
        print("\nModel Performance Comparison:")
        print(evaluation)
        
        # Step 7: Visualize results
        print("\n7. Visualizing results...")
        visualize_forecasts(enhanced_data, final_train_idx, final_test_idx, 
                         forecasts, model_names, evaluation)
        
        return {
            'evaluation': evaluation,
            'models': {
                'sarima': sarima_model,
                'lstm': lstm_model,
                'xgboost': xgb_models
            },
            'forecasts': dict(zip(model_names, forecasts)),
            'data': enhanced_data,
            'train_idx': final_train_idx,
            'test_idx': final_test_idx
        }
    
    except Exception as e:
        print(f"Error in forecasting pipeline: {str(e)}")
        print("Stack trace:")
        import traceback
        traceback.print_exc()
        return None

###############################################################################
# EXECUTE IF RUN AS SCRIPT
###############################################################################

if __name__ == "__main__":
    # Path to the data file
    data_file = "https://fetch-hiring.s3.amazonaws.com/machine-learning-engineer/receipt-count-prediction/data_daily.csv"
    
    # Run the complete pipeline
    results = run_complete_forecasting_pipeline(data_file)
    
    # The results dictionary contains everything you need to further analyze or use the models
    print("\nForecasting pipeline completed successfully!")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss, acf, pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import xgboost as xgb
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from scipy.signal import find_peaks
import warnings
from pathlib import Path

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')
tf.keras.utils.set_random_seed(42)  # For reproducibility

###############################################################################
# DATA LOADING AND PREPROCESSING
###############################################################################

def load_data(file_path):
    """Load time series data from a CSV file"""
    df = pd.read_csv(file_path, parse_dates=['# Date'])
    
    # Rename date column
    df = df.rename(columns={'# Date': 'Date'})
    
    # Set the date column as the index
    df = df.set_index('Date')
    
    # Ensure the data is sorted by date
    df = df.sort_index()
    
    # Do NOT convert to period index since it causes plotting issues
    # Instead ensure it's a DatetimeIndex
    if not isinstance(df.index, pd.DatetimeIndex):
        df.index = pd.to_datetime(df.index)
    
    return df

###############################################################################
# FEATURE ENGINEERING
###############################################################################

def enhance_features(df):
    """Enhanced feature engineering for time series data"""
    # Create a copy to avoid modifying the original
    df = df.copy()
    
    # Ensure proper datetime index
    if isinstance(df.index, pd.PeriodIndex):
        df.index = df.index.to_timestamp()
    elif not isinstance(df.index, pd.DatetimeIndex):
        try:
            df.index = pd.to_datetime(df.index)
        except:
            print("Warning: Could not convert index to DatetimeIndex")
    
    # Calendar features
    try:
        df['dayofweek'] = df.index.dayofweek
        df['day'] = df.index.day  # Day of month
        df['month'] = df.index.month
        df['quarter'] = df.index.quarter
        df['year'] = df.index.year
        df['dayofyear'] = df.index.dayofyear
        
        # Handle potential errors with isocalendar
        try:
            df['weekofyear'] = df.index.isocalendar().week.astype(int)
        except:
            # Fallback if isocalendar attribute error occurs
            df['weekofyear'] = df.index.week
        
        df['is_weekend'] = (df.index.dayofweek >= 5).astype(int)
        df['is_month_end'] = df.index.is_month_end.astype(int)
        df['is_month_start'] = df.index.is_month_start.astype(int)
        df['is_quarter_end'] = df.index.is_quarter_end.astype(int)
        df['is_quarter_start'] = df.index.is_quarter_start.astype(int)
    except Exception as e:
        print(f"Warning: Error creating calendar features: {str(e)}")
    
    # Try to add US holidays if the holidays package is available
    try:
        import holidays
        us_holidays = holidays.US()
        df['is_holiday'] = df.index.to_series().apply(lambda x: x in us_holidays).astype(int)
    except:
        print("Holidays package not available, skipping holiday feature")
    
    # Add day name for easier interpretation
    day_names = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 
                4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
    if 'dayofweek' in df.columns:
        df['dayname'] = df['dayofweek'].map(day_names)
    
    # Rolling statistics - careful about leakage!
    df['rolling_mean_7d'] = df['Receipt_Count'].rolling(window=7, min_periods=1).mean()
    df['rolling_std_7d'] = df['Receipt_Count'].rolling(window=7, min_periods=1).std()
    df['rolling_mean_28d'] = df['Receipt_Count'].rolling(window=28, min_periods=1).mean()
    
    # Lag features - be careful about leakage when using these!
    for lag in [1, 7, 14, 28]:
        df[f'lag_{lag}'] = df['Receipt_Count'].shift(lag)
    
    # Add exponential weighted moving averages (EWMAs)
    df['ewma_7d'] = df['Receipt_Count'].ewm(span=7).mean()
    df['ewma_14d'] = df['Receipt_Count'].ewm(span=14).mean()
    df['ewma_30d'] = df['Receipt_Count'].ewm(span=30).mean()
    
    # Add cyclical encoding for temporal features
    df['dayofweek_sin'] = np.sin(2 * np.pi * df['dayofweek'] / 7)
    df['dayofweek_cos'] = np.cos(2 * np.pi * df['dayofweek'] / 7)
    df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
    df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
    df['dayofyear_sin'] = np.sin(2 * np.pi * df['dayofyear'] / 365.25)
    df['dayofyear_cos'] = np.cos(2 * np.pi * df['dayofyear'] / 365.25)
    
    # Add trend indicators
    df['trend_7d'] = df['Receipt_Count'].diff(7)
    df['trend_14d'] = df['Receipt_Count'].diff(14)
    
    # Add ratio features
    df['ratio_7_28'] = df['rolling_mean_7d'] / df['rolling_mean_28d']
    
    # Handle missing values from rolling/lag operations
    df = df.fillna(method='bfill').fillna(method='ffill')
    
    return df

###############################################################################
# TIME SERIES ANALYSIS
###############################################################################

def analyze_time_series(data, plot=True):
    """Improved analysis for stationarity and seasonality"""
    # Ensure we're working with a Series
    if isinstance(data, pd.DataFrame):
        data = data['Receipt_Count']
    
    print("Time Series Analysis:")
    print(f"Data points: {len(data)}")
    print(f"Start date: {data.index[0]}")
    print(f"End date: {data.index[-1]}")
    print(f"Min value: {data.min():.2f}, Max value: {data.max():.2f}")
    print(f"Mean: {data.mean():.2f}, Median: {data.median():.2f}")
    
    # Test for stationarity with ADF test
    adf_result = adfuller(data.dropna())
    print("\nAugmented Dickey-Fuller Test:")
    print(f"ADF Statistic: {adf_result[0]:.4f}")
    print(f"p-value: {adf_result[1]:.4f}")
    print(f"Critical Values: {adf_result[4]}")
    is_stationary = adf_result[1] < 0.05
    print(f"Is stationary: {is_stationary}")
    
    # Determine seasonality using autocorrelation
    acf_values = acf(data.dropna(), nlags=365, fft=True)
    
    # Find peaks in ACF
    peaks, _ = find_peaks(acf_values)
    
    seasonal_periods = []
    if len(peaks) > 0:
        # Find significant peaks
        for peak in peaks:
            if peak > 1 and acf_values[peak] > 0.2:  # Threshold for significance
                seasonal_periods.append(peak)
                if len(seasonal_periods) >= 3:
                    break
        
        print("\nPotential seasonal periods detected in ACF:")
        for period in seasonal_periods:
            print(f"- Period of {period} days (correlation: {acf_values[period]:.3f})")
    
    # Check for weekly, monthly, and yearly patterns
    seasonal_results = {}
    for period in [7, 30, 90, 365]:  # Weekly, monthly, quarterly, yearly
        try:
            result = seasonal_decompose(data, model='additive', period=period)
            seasonal_variance = np.var(result.seasonal)
            seasonal_results[period] = seasonal_variance
            print(f"Seasonal variance with period={period}: {seasonal_variance:.4f}")
        except:
            print(f"Could not decompose with period={period}")
    
    # Select strongest seasonal period
    best_period = 7  # Default weekly
    if seasonal_results:
        best_period = max(seasonal_results.items(), key=lambda x: x[1])[0]
        print(f"Best seasonal period based on variance: {best_period}")
    
    # Plot the decomposition using the best period
    if plot and seasonal_results:
        plt.figure(figsize=(15, 12))
        result = seasonal_decompose(data, model='additive', period=best_period)
        result.plot()
        plt.tight_layout()
        plt.show()
        
        # Plot autocorrelation function
        plt.figure(figsize=(15, 6))
        plot_acf(data, lags=50, title="Autocorrelation Function")
        plt.grid(True)
        plt.show()
        
        # Plot partial autocorrelation function
        plt.figure(figsize=(15, 6))
        plot_pacf(data, lags=50, title="Partial Autocorrelation Function")
        plt.grid(True)
        plt.show()
    
    return {
        "is_stationary": is_stationary, 
        "seasonal_period": best_period, 
        "detected_periods": seasonal_periods,
        "seasonal_variances": seasonal_results
    }

###############################################################################
# TIME SERIES CROSS-VALIDATION
###############################################################################

def time_series_cv_split(data, n_splits=5, test_size=0.2):
    """Create properly separated time series folds"""
    # Calculate the size of each fold
    fold_size = int(len(data) / (n_splits + 1))  # +1 for final evaluation
    
    cv_splits = []
    for i in range(n_splits):
        # Calculate start and end points for each fold
        train_end = (i + 1) * fold_size
        test_start = train_end
        test_end = test_start + fold_size
        
        # Safety checks for indices
        test_end = min(test_end, len(data))
        
        # Create train/test indices
        train_indices = list(range(0, train_end))
        test_indices = list(range(test_start, test_end))
        
        cv_splits.append((train_indices, test_indices))
    
    # Create final train/test split for evaluation
    train_size = int(len(data) * (1 - test_size))
    final_train = list(range(0, train_size))
    final_test = list(range(train_size, len(data)))
    
    return cv_splits, (final_train, final_test)

###############################################################################
# IMPROVED LSTM IMPLEMENTATION
###############################################################################

def prepare_multivariate_lstm_data(data, train_indices, test_indices, seq_length=30):
    """Properly prepare multivariate data for LSTM with careful handling of train/test split"""
    print("Preparing multivariate LSTM data...")
    
    # Define important features for LSTM
    selected_cols = [
        'Receipt_Count',               # Target variable
        'dayofweek', 'month', 'quarter', 'is_weekend',  # Calendar features
        'rolling_mean_7d', 'rolling_std_7d', 'rolling_mean_28d',  # Rolling stats
        'dayofweek_sin', 'dayofweek_cos', 'month_sin', 'month_cos',  # Cyclical encoding
        'ewma_7d', 'ewma_14d',         # Exponential moving averages
        'trend_7d', 'trend_14d',       # Trend indicators
    ]
    
    # Add more features if they exist
    additional_features = ['is_holiday', 'is_month_end', 'is_quarter_end']
    for feature in additional_features:
        if feature in data.columns:
            selected_cols.append(feature)
    
    # Select features and fill any missing values
    model_data = data[selected_cols].copy()
    model_data = model_data.fillna(method='bfill').fillna(method='ffill')
    
    # Scale data - Scale each feature independently
    scalers = {}
    scaled_data = pd.DataFrame(index=model_data.index)
    
    for col in model_data.columns:
        scaler = MinMaxScaler(feature_range=(0, 1))
        # Fit scaler only on training data to prevent data leakage
        scaler.fit(model_data.iloc[train_indices][[col]])
        scaled_data[col] = scaler.transform(model_data[[col]]).flatten()
        scalers[col] = scaler
    
    # Create sequences for train set
    X_train, y_train = [], []
    
    for i in range(len(train_indices) - seq_length):
        if i + seq_length < len(train_indices):  # Ensure we don't go beyond train indices
            X_train.append(scaled_data.iloc[train_indices[i:i+seq_length]][selected_cols].values)
            y_train.append(scaled_data.iloc[train_indices[i+seq_length]]['Receipt_Count'])
    
    # Create sequences for test set
    X_test, y_test = [], []
    
    # Use the last sequence_length days from training for the first prediction
    for i in range(len(test_indices)):
        # For each test point, we need the previous seq_length points as input
        # This could include both training and test points
        seq_end_idx = train_indices[-1] + i + 1
        seq_start_idx = seq_end_idx - seq_length
        
        # Ensure we have sufficient data
        if seq_start_idx >= 0 and seq_end_idx < len(data):
            # Get indices for the sequence
            seq_indices = list(range(seq_start_idx, seq_end_idx))
            X_test.append(scaled_data.iloc[seq_indices][selected_cols].values)
            
            # The target is the test point
            if i < len(test_indices):
                y_test.append(scaled_data.iloc[train_indices[-1] + i + 1]['Receipt_Count'])
    
    # Convert to numpy arrays
    X_train = np.array(X_train)
    y_train = np.array(y_train)
    X_test = np.array(X_test)
    y_test = np.array(y_test)
    
    print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
    print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
    
    # Create validation set from the training data
    val_size = int(len(X_train) * 0.2)
    X_val = X_train[-val_size:]
    y_val = y_train[-val_size:]
    X_train = X_train[:-val_size]
    y_train = y_train[:-val_size]
    
    print(f"After validation split:")
    print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
    print(f"X_val shape: {X_val.shape}, y_val shape: {y_val.shape}")
    
    return X_train, y_train, X_val, y_val, X_test, y_test, scalers, selected_cols

def create_advanced_lstm_model(input_shape):
    """Create an advanced LSTM model with bidirectional layers and attention mechanism"""
    print(f"Creating advanced LSTM model with input shape: {input_shape}")
    
    model = Sequential()
    
    # First Bidirectional LSTM layer with more units
    model.add(Bidirectional(
        LSTM(128, return_sequences=True, 
             activation='tanh', recurrent_activation='sigmoid',
             dropout=0.2, recurrent_dropout=0.1),
        input_shape=input_shape
    ))
    model.add(BatchNormalization())
    
    # Second Bidirectional LSTM layer
    model.add(Bidirectional(
        LSTM(64, return_sequences=False,
             activation='tanh', recurrent_activation='sigmoid',
             dropout=0.2, recurrent_dropout=0.1)
    ))
    model.add(BatchNormalization())
    model.add(Dropout(0.3))
    
    # Dense layers
    model.add(Dense(64, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    
    model.add(Dense(32, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    
    # Output layer
    model.add(Dense(1))
    
    # Compile with Huber loss for robustness to outliers
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
    model.compile(
        loss=tf.keras.losses.Huber(delta=1.0),
        optimizer=optimizer,
        metrics=['mse', 'mae']
    )
    
    # Print model summary
    model.summary()
    
    return model

def train_lstm_model(model, X_train, y_train, X_val, y_val, epochs=200, batch_size=32):
    """Train the LSTM model with proper callbacks and monitoring"""
    print("Training LSTM model...")
    
    # Callbacks for better training
    callbacks = [
        EarlyStopping(
            monitor='val_loss',
            patience=20,
            restore_best_weights=True,
            verbose=1
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=10,
            min_lr=0.00001,
            verbose=1
        ),
        ModelCheckpoint(
            'best_lstm_model.h5',
            monitor='val_loss',
            save_best_only=True,
            verbose=1
        )
    ]
    
    # Train the model
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=callbacks,
        verbose=1,
        shuffle=False  # Important for time series
    )
    
    # Plot training history
    plt.figure(figsize=(12, 6))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('LSTM Training History')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.yscale('log')  # Log scale for better visualization
    plt.show()
    
    # Try to load the best model
    try:
        best_model = tf.keras.models.load_model('best_lstm_model.h5')
        print("Loaded best model from checkpoint")
        return best_model, history
    except:
        print("Using final model (checkpoint loading failed)")
        return model, history

def generate_lstm_forecast(model, X_test, scalers, test_indices, data):
    """Generate forecasts from the LSTM model with proper inverse scaling"""
    print("Generating LSTM forecasts...")
    
    # Make predictions
    scaled_predictions = model.predict(X_test, verbose=1)
    
    # Inverse transform the predictions to the original scale
    receipt_count_scaler = scalers['Receipt_Count']
    predictions = receipt_count_scaler.inverse_transform(scaled_predictions.reshape(-1, 1)).flatten()
    
    # Ensure predictions match the length of test_indices
    if len(predictions) > len(test_indices):
        predictions = predictions[:len(test_indices)]
    elif len(predictions) < len(test_indices):
        # Pad with the last prediction if needed
        padding = np.full(len(test_indices) - len(predictions), predictions[-1] if len(predictions) > 0 else 0)
        predictions = np.concatenate([predictions, padding])
    
    print(f"Generated {len(predictions)} LSTM forecasts")
    return predictions

###############################################################################
# SARIMA MODEL
###############################################################################

def fit_improved_sarima(data, seasonal_analysis):
    """Improved SARIMA with proper parameter selection"""
    # Get seasonal period from analysis
    seasonal_period = seasonal_analysis["seasonal_period"]
    
    print(f"Fitting SARIMA with seasonal period: {seasonal_period}")
    
    # Try to use auto_arima if available, otherwise use fixed parameters
    try:
        from pmdarima import auto_arima
        
        print("Using auto_arima to find best parameters...")
        auto_model = auto_arima(
            data,
            seasonal=True,
            m=seasonal_period,
            start_p=0, max_p=3,
            start_q=0, max_q=3,
            start_P=0, max_P=2,
            start_Q=0, max_Q=2,
            d=None, D=None,  # Allow auto_arima to determine
            trace=True,
            error_action='ignore',
            suppress_warnings=True,
            stepwise=True
        )
        
        order = auto_model.order
        seasonal_order = auto_model.seasonal_order
        print(f"Best SARIMA parameters: {order} {seasonal_order}")
        
    except ImportError:
        # Fallback to standard parameters if auto_arima is not available
        print("auto_arima not available, using default parameters")
        order = (1, 1, 1)
        seasonal_order = (1, 1, 1, seasonal_period)
    
    # Fit final model with best parameters
    print("Fitting final SARIMA model...")
    final_model = sm.tsa.statespace.SARIMAX(
        data,
        order=order,
        seasonal_order=seasonal_order,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    
    final_model_fit = final_model.fit(disp=False)
    print(f"SARIMA model AIC: {final_model_fit.aic:.2f}")
    
    return final_model_fit

###############################################################################
# IMPROVED XGBOOST
###############################################################################

def improved_xgboost(data, ts_analysis, cv_splits):
    """XGBoost with improved features and hyperparameter tuning"""
    print("Training XGBoost model with improved features...")
    
    # Create advanced features specifically for XGBoost
    data = data.copy()
    
    # Add exponential weighted moving averages
    data['ewma_7d'] = data['Receipt_Count'].ewm(span=7).mean()
    data['ewma_14d'] = data['Receipt_Count'].ewm(span=14).mean()
    data['ewma_30d'] = data['Receipt_Count'].ewm(span=30).mean()
    
    # Add rolling statistics
    for window in [7, 14, 30, 90]:
        data[f'roll_mean_{window}d'] = data['Receipt_Count'].rolling(window=window, min_periods=1).mean()
        data[f'roll_std_{window}d'] = data['Receipt_Count'].rolling(window=window, min_periods=1).std()
        data[f'roll_max_{window}d'] = data['Receipt_Count'].rolling(window=window, min_periods=1).max()
        data[f'roll_min_{window}d'] = data['Receipt_Count'].rolling(window=window, min_periods=1).min()
    
    # Add trend indicators
    data['trend_7d'] = data['Receipt_Count'].diff(7)
    data['trend_14d'] = data['Receipt_Count'].diff(14)
    data['trend_30d'] = data['Receipt_Count'].diff(30)
    
    # Add day of year cyclical features using sine and cosine transforms
    data['dayofyear_sin'] = np.sin(2 * np.pi * data.index.dayofyear/365.25)
    data['dayofyear_cos'] = np.cos(2 * np.pi * data.index.dayofyear/365.25)
    
    # Add month cyclical features
    data['month_sin'] = np.sin(2 * np.pi * data.index.month/12)
    data['month_cos'] = np.cos(2 * np.pi * data.index.month/12)
    
    # Add day of week cyclical features
    data['dayofweek_sin'] = np.sin(2 * np.pi * data.index.dayofweek/7)
    data['dayofweek_cos'] = np.cos(2 * np.pi * data.index.dayofweek/7)
    
    # Add ratio features
    data['ratio_7_30'] = data['roll_mean_7d'] / data['roll_mean_30d']
    data['ratio_7_90'] = data['roll_mean_7d'] / data['roll_mean_90d']
    
    # Fill NaN values
    data = data.fillna(method='bfill').fillna(method='ffill')
    
    # Prepare feature columns (exclude the target)
    feature_cols = [col for col in data.columns if col != 'Receipt_Count']
    
    # Create feature sets for each fold to prevent leakage
    fold_models = []
    fold_metrics = []
    
    # Store feature importances
    all_importances = {}
    
    for fold_idx, (train_idx, test_idx) in enumerate(cv_splits):
        print(f"\nTraining XGBoost model for fold {fold_idx+1}/{len(cv_splits)}...")
        
        # Split data
        train_data = data.iloc[train_idx]
        test_data = data.iloc[test_idx]
        
        # Define X and y
        X_train = train_data[feature_cols]
        y_train = train_data['Receipt_Count']
        X_test = test_data[feature_cols]
        y_test = test_data['Receipt_Count']
        
        # Train XGBoost model with optimized parameters
        model = xgb.XGBRegressor(
            objective='reg:squarederror',
            max_depth=6,              # Increased depth for more complex patterns
            learning_rate=0.05,       # Lower learning rate for better convergence
            n_estimators=500,         # More estimators for better performance
            subsample=0.8,
            colsample_bytree=0.8,
            min_child_weight=3,       # Controls model complexity
            gamma=0.1,                # Minimum loss reduction to create a new tree split
            reg_alpha=0.1,            # L1 regularization
            reg_lambda=1.0,           # L2 regularization
            random_state=42
        )
        
        # Fit model - handle compatibility issues with different XGBoost versions
        try:
            # Try with early_stopping_rounds parameter
            eval_set = [(X_test, y_test)]
            model.fit(X_train, y_train, eval_set=eval_set, early_stopping_rounds=30, verbose=False)
        except TypeError:
            try:
                # Try with eval_metric parameter
                model.fit(X_train, y_train, eval_set=eval_set, eval_metric='rmse', verbose=False)
            except:
                # Fallback to basic fit
                model.fit(X_train, y_train)
        
        # Evaluate
        y_pred = model.predict(X_test)
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        mae = mean_absolute_error(y_test, y_pred)
        
        print(f"  Fold {fold_idx+1} - RMSE: {rmse:.2f}, MAE: {mae:.2f}")
        
        # Add to fold results
        fold_models.append(model)
        fold_metrics.append(rmse)
        
        # Store feature importances
        for feature, importance in zip(feature_cols, model.feature_importances_):
            if feature not in all_importances:
                all_importances[feature] = []
            all_importances[feature].append(importance)
    
    # Report average performance
    avg_rmse = np.mean(fold_metrics)
    std_rmse = np.std(fold_metrics)
    print(f"XGBoost CV RMSE: {avg_rmse:.2f} ± {std_rmse:.2f}")
    
    # Create a feature importance summary if we have results
    if all_importances:
        feature_importance = {feature: np.mean(importances) for feature, importances in all_importances.items()}
        importance_df = pd.DataFrame({
            'Feature': list(feature_importance.keys()),
            'Importance': list(feature_importance.values())
        }).sort_values('Importance', ascending=False)
        
        try:
            plt.figure(figsize=(12, 8))
            sns.barplot(x='Importance', y='Feature', data=importance_df.head(15))
            plt.title('XGBoost Feature Importance (Top 15)')
            plt.tight_layout()
            plt.show()
        except Exception as e:
            print(f"Couldn't plot feature importance: {str(e)}")
    
    # Return the ensemble of models
    return fold_models

###############################################################################
# EVALUATION FUNCTIONS
###############################################################################

def evaluate_forecasts(y_true, forecasts, model_names):
    """Comprehensive evaluation with multiple metrics"""
    metrics = {}
    
    for name, forecast in zip(model_names, forecasts):
        # Standard metrics
        rmse = np.sqrt(mean_squared_error(y_true, forecast))
        mae = mean_absolute_error(y_true, forecast)
        r2 = r2_score(y_true, forecast)
        
        # Direction accuracy (for trends)
        direction_true = np.diff(y_true) > 0
        direction_pred = np.diff(forecast) > 0
        direction_accuracy = np.mean(direction_true == direction_pred)
        
        # MAPE (handle zeros carefully)
        mask = y_true != 0
        if any(mask):
            mape = np.mean(np.abs((y_true[mask] - forecast[mask]) / y_true[mask])) * 100
        else:
            mape = np.nan
        
        # Store metrics
        metrics[name] = {
            'RMSE': rmse,
            'MAE': mae,
            'R²': r2,
            'Direction Accuracy': direction_accuracy,
            'MAPE': mape
        }
    
    # Create comparison dataframe
    metrics_df = pd.DataFrame(metrics).T
    
    # Calculate ranks
    for col in metrics_df.columns:
        # Lower is better for error metrics, higher better for R² and direction
        if col in ['R²', 'Direction Accuracy']:
            metrics_df[f'{col} Rank'] = metrics_df[col].rank(ascending=False)
        else:
            metrics_df[f'{col} Rank'] = metrics_df[col].rank()
    
    # Add average rank
    rank_cols = [col for col in metrics_df.columns if 'Rank' in col]
    metrics_df['Avg Rank'] = metrics_df[rank_cols].mean(axis=1)
    
    # Sort by average rank
    metrics_df = metrics_df.sort_values('Avg Rank')
    
    return metrics_df

###############################################################################
# VISUALIZATION FUNCTIONS
###############################################################################

def visualize_forecasts(data, train_idx, test_idx, forecasts, model_names, metrics_df):
    """Visualize time series forecasts with comprehensive plots"""
    try:
        # Extract train and test data
        train_data = data.iloc[train_idx]['Receipt_Count']
        test_data = data.iloc[test_idx]['Receipt_Count']
        
        # Ensure datetime index
        if isinstance(train_data.index, pd.PeriodIndex):
            train_data.index = train_data.index.to_timestamp()
        if isinstance(test_data.index, pd.PeriodIndex):
            test_data.index = test_data.index.to_timestamp()
        
        # Create figure
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12), 
                                      gridspec_kw={'height_ratios': [3, 1]})
        
        # Plot original data
        ax1.plot(train_data.index, train_data.values, label='Training Data', color='blue')
        ax1.plot(test_data.index, test_data.values, label='Test Data', color='green')
        
        # Define colors for each model with distinct colors
        colors = {
            'Naive': 'gray',
            'Seasonal Naive': 'silver',
            'SARIMA': 'purple',
            'LSTM': 'orange',
            'XGBoost': 'red'
        }
        
        # Store prediction errors for the error plot
        prediction_errors = {}
        
        # Plot forecasts with slightly different line styles for better visibility
        linestyles = ['-', '--', '-.', ':', (0, (3, 1, 1, 1))]
        
        for i, (name, forecast) in enumerate(zip(model_names, forecasts)):
            color = colors.get(name, 'black')
            linestyle = linestyles[i % len(linestyles)]
            
            ax1.plot(test_data.index, forecast, 
                    label=f'{name} Forecast', 
                    color=color,
                    linestyle=linestyle, 
                    linewidth=2)
            
            # Calculate errors for the error plot
            errors = test_data.values - forecast
            prediction_errors[name] = errors
        
        # Plot prediction errors in the second subplot
        for i, (name, errors) in enumerate(prediction_errors.items()):
            color = colors.get(name, 'black')
            linestyle = linestyles[i % len(linestyles)]
            
            ax2.plot(test_data.index, errors, 
                    label=f'{name} Error', 
                    color=color,
                    linestyle=linestyle,
                    alpha=0.7)
        
        # Add vertical line for train-test split
        split_date = train_data.index[-1]
        ax1.axvline(x=split_date, color='black', linestyle='--')
        ax2.axvline(x=split_date, color='black', linestyle='--')
        
        # Add horizontal line at zero for the error plot
        ax2.axhline(y=0, color='black', linestyle='-', alpha=0.3)
        
        # Format the x-axis to show dates nicely
        for ax in [ax1, ax2]:
            # Simplified date formatting to avoid conversion errors
            ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
            ax.xaxis.set_major_locator(mdates.MonthLocator())
            plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)
            ax.grid(True, alpha=0.3)
        
        # Add titles and labels
        ax1.set_title('Time Series Model Comparison', fontsize=16)
        ax1.set_ylabel('Receipt Count', fontsize=12)
        ax1.legend(loc='upper left', frameon=True, framealpha=0.8, fontsize=10)
        
        ax2.set_title('Prediction Errors', fontsize=14)
        ax2.set_xlabel('Date', fontsize=12)
        ax2.set_ylabel('Error', fontsize=12)
        ax2.legend(loc='upper left', frameon=True, framealpha=0.8, fontsize=10)
        
        # Add a text box with model metrics in the main plot
        metrics_text = "Test Set Metrics:\n"
        for name in model_names:
            rmse = metrics_df.loc[name, 'RMSE']
            mae = metrics_df.loc[name, 'MAE']
            metrics_text += f"{name}: RMSE={rmse:.0f}, MAE={mae:.0f}\n"
        
        # Position the text box in the upper right of the plot
        props = dict(boxstyle='round', facecolor='white', alpha=0.8)
        ax1.text(0.02, 0.02, metrics_text, transform=ax1.transAxes, fontsize=10,
                verticalalignment='bottom', horizontalalignment='left', bbox=props)
        
        plt.tight_layout()
        plt.show()
        
        # Also plot a metrics comparison chart
        try:
            plt.figure(figsize=(12, 8))
            metrics_plot = metrics_df[['RMSE', 'MAE', 'R²', 'Direction Accuracy']].copy()
            
            # Normalize for plotting
            for col in metrics_plot.columns:
                if col in ['R²', 'Direction Accuracy']:
                    # Higher is better, scale to 0-1
                    max_val = metrics_plot[col].max()
                    if max_val != 0:  # Avoid division by zero
                        metrics_plot[col] = metrics_plot[col] / max_val
                else:
                    # Lower is better, invert and scale
                    max_val = metrics_plot[col].max()
                    if max_val != 0:  # Avoid division by zero
                        metrics_plot[col] = 1 - (metrics_plot[col] / max_val)
            
            # Plot bar chart
            metrics_plot = metrics_plot.reset_index()
            metrics_plot = pd.melt(metrics_plot, id_vars='index', var_name='Metric', value_name='Score')
            metrics_plot.rename(columns={'index': 'Model'}, inplace=True)
            
            plt.figure(figsize=(12, 8))
            sns.barplot(x='Model', y='Score', hue='Metric', data=metrics_plot)
            plt.title('Model Comparison (Higher is Better)')
            plt.ylim(0, 1)
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()
        except Exception as e:
            print(f"Warning: Could not create metrics comparison chart: {str(e)}")
    
    except Exception as e:
        print(f"Warning: Could not create visualization: {str(e)}")
        print("Continuing without visualization...")

###############################################################################
# MAIN EXECUTION PIPELINE
###############################################################################

def run_complete_forecasting_pipeline(data_file):
    """Complete time series forecasting pipeline with all fixes"""
    try:
        # Step 1: Load and preprocess data
        print("\n1. Loading data...")
        data = load_data(data_file)
        print(f"Loaded {len(data)} rows of data")
        
        # Step 2: Analyze the time series
        print("\n2. Analyzing time series...")
        ts_analysis = analyze_time_series(data['Receipt_Count'])
        
        # Step 3: Enhance features
        print("\n3. Enhancing features...")
        enhanced_data = enhance_features(data.copy())
        print(f"Created {len(enhanced_data.columns)} features")
        
        # Step 4: Create proper time series CV splits
        print("\n4. Creating cross-validation splits...")
        cv_splits, (final_train_idx, final_test_idx) = time_series_cv_split(
            enhanced_data, n_splits=5, test_size=0.2
        )
        print(f"Created {len(cv_splits)} CV folds plus a final evaluation split")
        print(f"Final train size: {len(final_train_idx)}, test size: {len(final_test_idx)}")
        
        # Step 5: Train and evaluate models
        print("\n5. Training and evaluating models...")
        
        # 5.1 Baseline models
        print("\n5.1 Creating baseline forecasts...")
        naive_forecast = enhanced_data['Receipt_Count'].iloc[final_train_idx[-1]] * np.ones(len(final_test_idx))
        
        seasonal_naive_forecast = np.array([])
        period = ts_analysis['seasonal_period']
        for i in range(len(final_test_idx)):
            season_idx = (final_train_idx[-1] - period + i % period) % len(enhanced_data)
            seasonal_naive_forecast = np.append(seasonal_naive_forecast, 
                                            enhanced_data['Receipt_Count'].iloc[season_idx])
        
        # 5.2 SARIMA model
        print("\n5.2 Training SARIMA model...")
        try:
            sarima_model = fit_improved_sarima(
                enhanced_data['Receipt_Count'].iloc[final_train_idx], ts_analysis
            )
            sarima_forecast = sarima_model.forecast(steps=len(final_test_idx))
        except Exception as e:
            print(f"Error in SARIMA modeling: {str(e)}")
            # Fallback to naive forecast if SARIMA fails
            sarima_model = None
            sarima_forecast = naive_forecast.copy()
        
        # 5.3 LSTM model - COMPLETELY REWRITTEN
        print("\n5.3 Training LSTM model...")
        try:
            # Prepare data for LSTM using the multivariate approach
            X_train, y_train, X_val, y_val, X_test, y_test, scalers, feature_cols = prepare_multivariate_lstm_data(
                enhanced_data, final_train_idx, final_test_idx, seq_length=30
            )
            
            # Create and train the LSTM model
            lstm_model = create_advanced_lstm_model(input_shape=(X_train.shape[1], X_train.shape[2]))
            lstm_model, history = train_lstm_model(
                lstm_model, X_train, y_train, X_val, y_val, epochs=200, batch_size=32
            )
            
            # Generate LSTM forecasts
            lstm_forecast = generate_lstm_forecast(
                lstm_model, X_test, scalers, final_test_idx, enhanced_data
            )
            
        except Exception as e:
            print(f"Error in LSTM modeling: {str(e)}")
            import traceback
            traceback.print_exc()
            # Fallback to naive forecast if LSTM fails
            lstm_model = None
            lstm_forecast = naive_forecast.copy()
        
        # 5.4 XGBoost model
        print("\n5.4 Training XGBoost model...")
        try:
            xgb_models = improved_xgboost(enhanced_data, ts_analysis, cv_splits)
            
            # Prepare test features
            X_test_xgb = enhanced_data.iloc[final_test_idx].copy()
            feature_cols = [col for col in enhanced_data.columns if col != 'Receipt_Count']
            
            # Fix lag features to avoid data leakage
            for lag in [1, 7, 14, 28]:
                lag_col = f'lag_{lag}'
                if lag_col in X_test_xgb.columns:
                    for i, idx in enumerate(final_test_idx):
                        if idx - lag >= 0:
                            X_test_xgb.iloc[i, X_test_xgb.columns.get_loc(lag_col)] = enhanced_data.iloc[idx - lag]['Receipt_Count']
            
            # Generate predictions from each model
            xgb_forecasts = []
            for model in xgb_models:
                try:
                    # Get the features this model was trained with
                    model_features = [f for f in feature_cols if f in X_test_xgb.columns]
                    xgb_forecasts.append(model.predict(X_test_xgb[model_features]))
                except Exception as e:
                    print(f"Warning: Error predicting with XGBoost model: {str(e)}")
            
            # Use the mean of all model predictions if we have any
            if xgb_forecasts:
                xgb_forecast = np.mean(xgb_forecasts, axis=0)
            else:
                xgb_forecast = naive_forecast.copy()
        except Exception as e:
            print(f"Error in XGBoost modeling: {str(e)}")
            # Fallback to naive forecast if XGBoost fails
            xgb_models = []
            xgb_forecast = naive_forecast.copy()
        
        # Step 6: Evaluate all models
        print("\n6. Evaluating all models...")
        y_true = enhanced_data['Receipt_Count'].iloc[final_test_idx]
        
        forecasts = [naive_forecast, seasonal_naive_forecast, sarima_forecast, lstm_forecast, xgb_forecast]
        model_names = ['Naive', 'Seasonal Naive', 'SARIMA', 'LSTM', 'XGBoost']
        
        # Ensure all forecasts are the same length as y_true
        for i in range(len(forecasts)):
            if len(forecasts[i]) != len(y_true):
                print(f"Warning: {model_names[i]} forecast length mismatch. Adjusting...")
                if len(forecasts[i]) > len(y_true):
                    forecasts[i] = forecasts[i][:len(y_true)]
                else:
                    # Pad with last value if too short
                    forecasts[i] = np.append(forecasts[i], 
                                           np.full(len(y_true) - len(forecasts[i]), forecasts[i][-1]))
        
        evaluation = evaluate_forecasts(y_true, forecasts, model_names)
        
        # Print detailed evaluation
        print("\nModel Performance Comparison:")
        print(evaluation)
        
        # Step 7: Visualize results
        print("\n7. Visualizing results...")
        visualize_forecasts(enhanced_data, final_train_idx, final_test_idx, 
                         forecasts, model_names, evaluation)
        
        return {
            'evaluation': evaluation,
            'models': {
                'sarima': sarima_model,
                'lstm': lstm_model,
                'xgboost': xgb_models
            },
            'forecasts': dict(zip(model_names, forecasts)),
            'data': enhanced_data,
            'train_idx': final_train_idx,
            'test_idx': final_test_idx
        }
    
    except Exception as e:
        print(f"Error in forecasting pipeline: {str(e)}")
        print("Stack trace:")
        import traceback
        traceback.print_exc()
        return None

###############################################################################
# EXECUTE IF RUN AS SCRIPT
###############################################################################

if __name__ == "__main__":
    # Path to the data file
    data_file = "https://fetch-hiring.s3.amazonaws.com/machine-learning-engineer/receipt-count-prediction/data_daily.csv"
    
    # Run the complete pipeline
    results = run_complete_forecasting_pipeline(data_file)
    
    # The results dictionary contains everything you need to further analyze or use the models
    print("\nForecasting pipeline completed successfully!")
```
