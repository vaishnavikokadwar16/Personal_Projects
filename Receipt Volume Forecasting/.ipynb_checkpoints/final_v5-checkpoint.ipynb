```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
seasonal_results = {}
for period in [7, 30, 60, 90, 365]:  # Weekly, monthly, quarterly, yearly
    try:
        result = seasonal_decompose(df['Receipt_Count'], model='additive', period=period)
        seasonal_variance = np.var(result.seasonal)
        seasonal_results[period] = seasonal_variance
        print(f"Seasonal variance with period={period}: {seasonal_variance:.4f}")
    except:
        print(f"Could not decompose with period={period}")

# Select strongest seasonal period
best_period = 7  # Default weekly
if seasonal_results:
    best_period = max(seasonal_results.items(), key=lambda x: x[1])[0]
    print(f"Best seasonal period based on variance: {best_period}")
```

```python
def mape_calculation(test, forecast_values, title):
    # Extract the actual and forecast values from the DataFrame
    actual_values = test['Receipt_Count'].values
    forecast_values = forecast_values
    
    # Ensure both are one-dimensional arrays
    actual_values = np.squeeze(actual_values)
    forecast_values = np.squeeze(forecast_values)
    
    # Calculate the MAPE for each row
    mape_row_by_row = np.abs((actual_values - forecast_values) / actual_values) * 100
    
    # Create a DataFrame to display the row-by-row MAPE along with the actual and forecast values
    df_results = test.copy()
    df_results['Actual'] = actual_values
    df_results['Forecast'] = forecast_values
    df_results['MAPE'] = mape_row_by_row
    
    # Calculate the overall weighted MAPE
    weights = actual_values / actual_values.sum()
    overall_mape = np.sum(mape_row_by_row * weights)
    
    # Print the overall MAPE
    print(f"Overall MAPE: {overall_mape:.2f}%")

    rmse = np.sqrt(mean_squared_error(actual_values, forecast_values))
    mae = mean_absolute_error(actual_values, forecast_values)
    r2 = r2_score(actual_values, forecast_values)

    print(f"Overall RMSE: {rmse:.2f}")
    print(f"Overall MAE: {mae:.2f}")
    print(f"Overall R2: {r2:.2f}")

    # Direction accuracy (for trends)
    direction_true = np.diff(actual_values) > 0
    direction_pred = np.diff(forecast_values) > 0
    direction_accuracy = np.mean(direction_true == direction_pred)

    print(f"Overall Directional Accuracy: {direction_accuracy:.2f}%")

    #title = 'Auto ARIMA with 3 Exogenous Variables'
    ylabel='Receipts Count'
    
    plt.figure(figsize=(12, 6))
    plt.plot(df_results['Actual'], label='Observed Data')
    plt.plot(df_results['Forecast'], label='Forecasted Data', linestyle='--')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Receipt Count')
    plt.legend()
    plt.show()
    
    plt.show();
```

```python
# Scale the data
scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))

X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

y_train_scaled = scaler_y.fit_transform(np.array(y_train).reshape(-1, 1))
y_test_scaled = scaler_y.transform(np.array(y_test).reshape(-1, 1))

# Convert to PyTorch tensors
X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32).unsqueeze(1)
X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32).unsqueeze(1)
y_train_tensor = torch.tensor(y_train_scaled, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test_scaled, dtype=torch.float32)

# Create DataLoader
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
```

```python
# Define the LSTM model
class LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, num_layers):
        super(LSTMModel, self).__init__()
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
    
    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim).to(x.device)
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out
```

```python
# Hyperparameters
input_dim = X_train_tensor.shape[2]
hidden_dim = 50
output_dim = 1
num_layers = 2
num_epochs = 100
learning_rate = 0.001

# Instantiate the model, define the loss function and the optimizer
model_v6 = LSTMModel(input_dim, hidden_dim, output_dim, num_layers)
criterion = nn.MSELoss()
optimizer = optim.Adam(model_v6.parameters(), lr=learning_rate)
```

```python
model_v6.train()
for epoch in range(num_epochs):
    for i, (x_batch, y_batch) in enumerate(train_loader):
        outputs = model_v6(x_batch)
        loss = criterion(outputs, y_batch)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python
model_v6.eval()
with torch.no_grad():
    predictions_v6 = model_v6(X_test_tensor).cpu().numpy()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
import matplotlib.pyplot as plt

df = df[['Receipt_Count']]
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df)

# Hyperparameters
sequence_length = 90  # Using last 30 days to predict the next day
epochs = 20
batch_size = 32

# Prepare the data for RNN
def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i + sequence_length, 0])
        y.append(data[i + sequence_length, 0])
    return np.array(X), np.array(y)

X, y = create_sequences(scaled_data, sequence_length)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# Split data into training and testing sets
train_size = len(X) - 30
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build the RNN model
model = Sequential()
model.add(SimpleRNN(50, input_shape=(sequence_length, 1), activation='tanh', return_sequences=False))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.1, verbose=1)

# Make predictions
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)

# Inverse scaling
train_pred = scaler.inverse_transform(train_pred.reshape(-1, 1))
test_pred = scaler.inverse_transform(test_pred.reshape(-1, 1))
y_train = scaler.inverse_transform(y_train.reshape(-1, 1))
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

forecast_values_v7 = test_pred
```
