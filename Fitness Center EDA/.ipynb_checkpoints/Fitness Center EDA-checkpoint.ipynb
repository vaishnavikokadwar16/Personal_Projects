```python
# Group by 'price' and 'churn', calculate counts, and unstack 'churn' to get counts in separate columns
df_counts = df.groupby(['price', 'default']).size().unstack(fill_value=0)

# Calculate % churn and reset index to keep 'price' as a column
df_counts['% churn'] = df_counts[1] / (df_counts[0] + df_counts[1])*100
df_counts = df_counts.reset_index()

# Plot the results
sns.barplot(x='price', y='% churn', data=df_counts)

plt.show()
```

```python
plt.figure(figsize=(10, 4))

# First subplot
plt.subplot(1, 2, 1)
a = df[df['churn']==1]['monthdue'].value_counts().sort_index().to_numpy()  # Sort by the index to preserve original order
plt.plot(a, '.', markersize=2, color='C3')
plt.grid(linestyle=':')
plt.xlabel('Month Due')
plt.ylabel('Number of Drops per Month Due')

# Second subplot
plt.subplot(122)
plt.plot(a.cumsum() / a.sum() * 100, linewidth=2, color='C0')
ax = plt.gca()
ax.grid(linestyle=':')
ax.set_axisbelow(True)
ax.set_xlabel('Month Due')
ax.set_ylabel('Cumulative % of Drops')

plt.suptitle(f'{a.sum():,} Fitness Centre Customer Drops')
plt.show()
```

```python
df_cleaned = pd.read_excel('Fitness Club Data V4-1.xlsx')
```

```python
df_cleaned
```

```python
df_cleaned = df_cleaned[df_cleaned['age'].between(17,79)]
df_cleaned = pd.get_dummies(df_cleaned,columns=['use','pmttype'],drop_first=True,prefix=['use','pmttype'])
df_cleaned.rename(columns={'default':'churn'},inplace=True)
df_cleaned
```

Logistic Regression

```python
y = df_cleaned['churn']
x = df_cleaned.drop(['churn','enrolldt'],axis = 1)
```

```python
y.head()
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42, stratify=y)
```

```python
Logmodel = LogisticRegression()

Logmodel.fit(X_train, y_train)
```

```python
preds = Logmodel.predict(X_test)
```

```python
print(classification_report(y_test,preds))
```

```python
print(confusion_matrix(y_test,preds))
```
